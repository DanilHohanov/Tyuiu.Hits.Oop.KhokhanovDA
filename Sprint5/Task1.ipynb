<h1 style="color:DodgerBlue">HashSet<T></h1>



`HashSet<T>` в C# — это коллекция, которая хранит уникальные элементы без определенного порядка. Она полезна, когда нужно быстро проверять наличие элемента или добавлять только уникальные элементы. Давайте рассмотрим, как с ней работать на примере классов `Transport`, `Car`, и `Bike`.

### Создание классов Transport, Car, и Bike

Сначала определим базовый класс `Transport` и его наследники `Car` и `Bike`:

```csharp
public class Transport
{
    public string Model { get; set; }
    public string Manufacturer { get; set; }

    public Transport(string model, string manufacturer)
    {
        Model = model;
        Manufacturer = manufacturer;
    }

    public override bool Equals(object obj)
    {
        // Простая проверка равенства базируется на модели и производителе
        if (obj is Transport transport)
        {
            return Model == transport.Model && Manufacturer == transport.Manufacturer;
        }
        return false;
    }

    public override int GetHashCode()
    {
        // Создание хеш-кода на основе модели и производителя
        return HashCode.Combine(Model, Manufacturer);
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Model: {Model}, Manufacturer: {Manufacturer}");
    }
}

public class Car : Transport
{
    public int NumberOfDoors { get; set; }

    public Car(string model, string manufacturer, int numberOfDoors)
        : base(model, manufacturer)
    {
        NumberOfDoors = numberOfDoors;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Number of Doors: {NumberOfDoors}");
    }
}

public class Bike : Transport
{
    public bool HasCarrier { get; set; }

    public Bike(string model, string manufacturer, bool hasCarrier)
        : base(model, manufacturer)
    {
        HasCarrier = hasCarrier;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Has Carrier: {HasCarrier}");
    }
}
```

### Работа с HashSet<T>

Теперь рассмотрим, как использовать `HashSet<T>` для хранения объектов `Transport`.

```csharp
using System;
using System.Collections.Generic;


// Создаем HashSet для уникальных объектов транспортных средств
HashSet<Transport> transportSet = new HashSet<Transport>();

// Добавляем объекты Car и Bike в HashSet
var car1 = new Car("Model S", "Tesla", 4);
var car2 = new Car("Model 3", "Tesla", 4);
 var bike1 = new Bike("Mountain Explorer", "Giant", true);

transportSet.Add(car1);
transportSet.Add(car2);
transportSet.Add(bike1);

// Попробуем добавить повторяющийся объект
var duplicateCar = new Car("Model S", "Tesla", 4);
bool isAdded = transportSet.Add(duplicateCar);

Console.WriteLine($"Was duplicate car added? {isAdded}"); // false

// Перебираем и выводим информацию обо всех уникальных транспортных средствах
foreach (var transport in transportSet)
{
    transport.DisplayInfo();
    Console.WriteLine();
}

```

### Объяснения и комментарии:

- **Уникальность объектов**: `HashSet<T>` не допускает дублирования. Если вы пытаетесь добавить элемент, который уже присутствует (определяется методом `Equals` и `GetHashCode`), он не добавляется.

- **Методы `Equals` и `GetHashCode`**: Эти методы переопределены для корректной работы с `HashSet`. `Equals` определяет, когда два объекта считаются равными, а `GetHashCode` возвращает хеш-код, используемый для быстрого поиска элемента.

- **Быстрая проверка на наличие**: Благодаря хешированию, `HashSet` позволяет быстро проверять, существует ли элемент в коллекции (`O(1)` — амортизированное время).

- **Отсутствие порядка**: Элементы в `HashSet` не имеют определенного порядка, в отличие от `List<T>`.

`HashSet<T>` идеально подходит для сценариев, где требуется исключительно уникальность элементов и быстрое выполнение операций поиска.

````
using System;
using System.Collections.Generic;

public class Transport
{
    public string Model { get; set; }
    public string Manufacturer { get; set; }

    public Transport(string model, string manufacturer)
    {
        Model = model;
        Manufacturer = manufacturer;
    }

    public override bool Equals(object obj)
    {
        // Простая проверка равенства базируется на модели и производителе
        if (obj is Transport transport)
        {
            return Model == transport.Model && Manufacturer == transport.Manufacturer;
        }
        return false;
    }

    public override int GetHashCode()
    {
        // Создание хеш-кода на основе модели и производителя
        return HashCode.Combine(Model, Manufacturer);
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Model: {Model}, Manufacturer: {Manufacturer}");
    }
}

public class Car : Transport
{
    public int NumberOfDoors { get; set; }

    public Car(string model, string manufacturer, int numberOfDoors)
        : base(model, manufacturer)
    {
        NumberOfDoors = numberOfDoors;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Number of Doors: {NumberOfDoors}");
    }
}

public class Bike : Transport
{
    public bool HasCarrier { get; set; }

    public Bike(string model, string manufacturer, bool hasCarrier)
        : base(model, manufacturer)
    {
        HasCarrier = hasCarrier;
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Has Carrier: {HasCarrier}");
    }
}

class Program
{
    static void Main()
    {
        // Создаем HashSet для уникальных объектов транспортных средств
        HashSet<Transport> transportSet = new HashSet<Transport>();

        // Добавляем объекты Car и Bike в HashSet
        var car1 = new Car("Model S", "Tesla", 4);
        var car2 = new Car("Model 3", "Tesla", 4);
        var bike1 = new Bike("Mountain Explorer", "Giant", true);

        transportSet.Add(car1);
        transportSet.Add(car2);
        transportSet.Add(bike1);

        Console.WriteLine("=== Начальный набор транспортных средств ===");
        foreach (var transport in transportSet)
        {
            transport.DisplayInfo();
            Console.WriteLine();
        }

        // Попробуем добавить повторяющийся объект
        Console.WriteLine("=== Попытка добавить дубликат ===");
        var duplicateCar = new Car("Model S", "Tesla", 4);
        bool isAdded = transportSet.Add(duplicateCar);
        Console.WriteLine($"Was duplicate car added? {isAdded}"); // false

        // Попробуем добавить уникальный объект
        Console.WriteLine("\n=== Добавление нового уникального объекта ===");
        var bike2 = new Bike("Road Bike", "Trek", false);
        isAdded = transportSet.Add(bike2);
        Console.WriteLine($"Was new bike added? {isAdded}"); // true

        // Перебираем и выводим информацию обо всех уникальных транспортных средствах
        Console.WriteLine("\n=== Все уникальные транспортные средства ===");
        foreach (var transport in transportSet)
        {
            transport.DisplayInfo();
            Console.WriteLine();
        }

        // Проверка наличия объекта в HashSet
        Console.WriteLine("=== Проверка наличия объектов ===");
        var searchCar = new Car("Model 3", "Tesla", 4);
        bool containsCar = transportSet.Contains(searchCar);
        Console.WriteLine($"Contains Model 3 Tesla? {containsCar}"); // true

        var searchBike = new Bike("Nonexistent", "Brand", true);
        bool containsBike = transportSet.Contains(searchBike);
        Console.WriteLine($"Contains Nonexistent Bike? {containsBike}"); // false

        // Удаление объекта
        Console.WriteLine("\n=== Удаление объекта ===");
        bool removed = transportSet.Remove(car2);
        Console.WriteLine($"Was car2 removed? {removed}"); // true
        Console.WriteLine($"Count after removal: {transportSet.Count}");

        // Отображение финального набора
        Console.WriteLine("\n=== Финальный набор ===");
        foreach (var transport in transportSet)
        {
            transport.DisplayInfo();
            Console.WriteLine();
        }
    }
}
````

Ниже в блоке по примеру создайте базовый класс Animal и производные классы (3-4 например Dog, Cat, Bird и так далее) реализуйте при помощи  HashSet. 

````
using System;
using System.Collections.Generic;

// Базовый класс Animal
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
    }

    // Переопределяем Equals для сравнения по имени и возрасту
    public override bool Equals(object obj)
    {
        if (obj is Animal animal)
        {
            return Name == animal.Name && Age == animal.Age;
        }
        return false;
    }

    // Переопределяем GetHashCode для корректной работы в HashSet
    public override int GetHashCode()
    {
        return HashCode.Combine(Name, Age);
    }

    public virtual void MakeSound()
    {
        Console.WriteLine($"{Name} издаёт звук...");
    }

    public virtual void DisplayInfo()
    {
        Console.WriteLine($"Имя: {Name}, Возраст: {Age} лет");
    }
}

// Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Порода: {Breed}");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик!");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу-мяу!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Домашний: {(IsIndoor ? "Да" : "Нет")}");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет...");
    }
}

// Производный класс Bird
public class Bird : Animal
{
    public double Wingspan { get; set; } // Размах крыльев в см

    public Bird(string name, int age, double wingspan) : base(name, age)
    {
        Wingspan = wingspan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Размах крыльев: {Wingspan} см");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летает!");
    }
}

// Производный класс Fish
public class Fish : Animal
{
    public string WaterType { get; set; } // "Пресная" или "Солёная"

    public Fish(string name, int age, string waterType) : base(name, age)
    {
        WaterType = waterType;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} булькает: Буль-буль!");
    }

    public override void DisplayInfo()
    {
        base.DisplayInfo();
        Console.WriteLine($"Тип воды: {WaterType}");
    }

    public void Swim()
    {
        Console.WriteLine($"{Name} плавает!");
    }
}

class Program
{
    static void Main()
    {
        // Создаем HashSet для уникальных объектов животных
        HashSet<Animal> animalSet = new HashSet<Animal>();

        Console.WriteLine("=== Добавление начальных животных ===");
        
        // Добавляем объекты различных типов животных
        var dog1 = new Dog("Рекс", 5, "Овчарка");
        var cat1 = new Cat("Мурка", 3, true);
        var bird1 = new Bird("Кеша", 2, 25.5);
        var fish1 = new Fish("Немо", 1, "Солёная");

        animalSet.Add(dog1);
        animalSet.Add(cat1);
        animalSet.Add(bird1);
        animalSet.Add(fish1);

        // Выводим все животные
        Console.WriteLine("\nНачальный набор животных:");
        foreach (var animal in animalSet)
        {
            animal.DisplayInfo();
            animal.MakeSound();
            Console.WriteLine(new string('-', 40));
        }

        // Попробуем добавить дубликат (то же имя и возраст)
        Console.WriteLine("\n=== Попытка добавить дубликат ===");
        var duplicateDog = new Dog("Рекс", 5, "Лабрадор"); // То же имя и возраст, но другая порода
        bool isAdded = animalSet.Add(duplicateDog);
        Console.WriteLine($"Был ли добавлен дубликат собаки? {isAdded}"); // false

        // Добавим уникальное животное
        Console.WriteLine("\n=== Добавление нового уникального животного ===");
        var cat2 = new Cat("Снежок", 4, false);
        isAdded = animalSet.Add(cat2);
        Console.WriteLine($"Был ли добавлен новый кот? {isAdded}"); // true

        // Проверка наличия объекта
        Console.WriteLine("\n=== Проверка наличия объектов ===");
        var searchCat = new Cat("Мурка", 3, true);
        bool containsCat = animalSet.Contains(searchCat);
        Console.WriteLine($"Содержится ли кот 'Мурка' возраста 3 года? {containsCat}"); // true

        var searchBird = new Bird("Попугай", 1, 20.0);
        bool containsBird = animalSet.Contains(searchBird);
        Console.WriteLine($"Содержится ли птица 'Попугай'? {containsBird}"); // false

        // Удаление объекта
        Console.WriteLine("\n=== Удаление объекта ===");
        bool removed = animalSet.Remove(bird1);
        Console.WriteLine($"Была ли удалена птица 'Кеша'? {removed}"); // true
        Console.WriteLine($"Количество животных после удаления: {animalSet.Count}");

        // Отображение финального набора
        Console.WriteLine("\n=== Финальный набор животных ===");
        foreach (var animal in animalSet)
        {
            animal.DisplayInfo();
            animal.MakeSound();
            
            // Вызов специфичных методов через проверку типа
            if (animal is Dog dog)
                dog.Fetch();
            else if (animal is Cat cat)
                cat.Purr();
            else if (animal is Fish fish)
                fish.Swim();
                
            Console.WriteLine(new string('-', 40));
        }

        // Демонстрация операций над множествами
        Console.WriteLine("\n=== Дополнительные операции с HashSet ===");
        
        // Создание второго множества
        HashSet<Animal> anotherSet = new HashSet<Animal>
        {
            new Dog("Тузик", 6, "Бультерьер"),
            new Cat("Мурка", 3, true), // Дубликат из первого множества
            new Bird("Воробей", 1, 15.0)
        };

        Console.WriteLine($"Первое множество: {animalSet.Count} животных");
        Console.WriteLine($"Второе множество: {anotherSet.Count} животных");

        // Объединение множеств
        animalSet.UnionWith(anotherSet);
        Console.WriteLine($"\nПосле объединения: {animalSet.Count} уникальных животных");

        // Пересечение множеств
        var intersection = new HashSet<Animal>(animalSet);
        intersection.IntersectWith(anotherSet);
        Console.WriteLine($"Общих животных во всех множествах: {intersection.Count}");
    }
}
````
