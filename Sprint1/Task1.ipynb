### Структура и Объявление Класса в C#

В C#, класс объявляется с использованием ключевого слова `class`, за которым следует имя класса. Класс может содержать поля (данные), свойства (способ доступа к данным), методы (функциональность) и события. 

#### Свойства

Свойства в C# используются для инкапсуляции данных. Они предоставляют гибкий механизм для чтения, записи или вычисления значений приватных полей. Свойства могут иметь два акцессора: `get` (геттер) для чтения значения и `set` (сеттер) для его изменения.

#### Геттеры и Сеттеры

Геттеры и сеттеры являются частью свойств класса в C#. Геттер (`get`) используется для возврата значения свойства, а сеттер (`set`) — для установки нового значения. Сеттер может включать логику валидации или другие операции перед изменением значения поля.

### Пример Классов Vehicle и Car

Давайте рассмотрим примеры классов `Vehicle` и `Car`, чтобы проиллюстрировать использование свойств, геттеров и сеттеров, а также разницу между статическими и обычными атрибутами и методами.

#### Класс Vehicle

```csharp
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}
```

#### Класс Car

```csharp
public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
```

### Разница между Статическими и Обычными Атрибутами и Методами Класса

- **Статические атрибуты и методы** принадлежат самому классу, а не его экземплярам. Они могут быть вызваны без создания объекта класса. В примере выше `TotalVehicles` является статическим атрибутом, а `DisplayTotalVehicles()` — статическим методом класса `Vehicle`.
- **Обычные атрибуты и методы** принадлежат экземплярам класса и могут иметь разные значения для разных объектов. В примере выше `_color`, `_speed` и `_fuelLevel` являются обычными атрибутами, а методы `Accelerate()` и `Refuel()` — обычными методами.

````
public class Vehicle
{
    private string _color;
    private int _speed;

    // Свойство Color с геттером и сеттером
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    // Свойство Speed с геттером и сеттером
    public int Speed
    {
        get { return _speed; }
        set
        {
            if (value >= 0)
                _speed = value;
            else
                throw new ArgumentOutOfRangeException("Скорость не может быть отрицательной!");
        }
    }

    // Статический атрибут
    public static int TotalVehicles { get; private set; } = 0;

    // Конструктор
    public Vehicle()
    {
        TotalVehicles++;
    }

    // Обычный метод
    public void Accelerate(int delta)
    {
        Speed += delta;
    }

    // Статический метод
    public static void DisplayTotalVehicles()
    {
        Console.WriteLine($"Общее количество транспортных средств: {TotalVehicles}");
    }
}

public class Car : Vehicle
{
    private int _fuelLevel;

    // Свойство FuelLevel с геттером и сеттером
    public int FuelLevel
    {
        get { return _fuelLevel; }
        set { _fuelLevel = value; }
    }

    // Конструктор
    public Car() : base()
    {
        FuelLevel = 100; // Задаем начальный уровень топлива
    }

    // Обычный метод
    public void Refuel(int amount)
    {
        FuelLevel += amount;
    }
}
````

Ниже в блоке по примеру создайте базовый класс Animal и производные классы (2-3 например Dog, Cat и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также статические и обычными атрибуты и метода класса

````
using System;

// Базовый класс Animal
public class Animal
{
    // Статический атрибут (принадлежит классу, а не экземплярам)
    public static int TotalAnimals { get; private set; }

    // Обычные атрибуты (поля)
    private string _name;
    private int _age;
    private double _weight;

    // Свойства с геттерами и сеттерами
    public string Name
    {
        get { return _name; }
        set 
        { 
            if (!string.IsNullOrEmpty(value))
                _name = value;
        }
    }

    public int Age
    {
        get { return _age; }
        set 
        { 
            if (value >= 0 && value <= 50)
                _age = value;
        }
    }

    public double Weight
    {
        get { return _weight; }
        set 
        { 
            if (value > 0)
                _weight = value;
        }
    }

    // Конструктор
    public Animal(string name, int age, double weight)
    {
        Name = name;
        Age = age;
        Weight = weight;
        TotalAnimals++; // Увеличиваем счетчик при создании животного
    }

    // Обычные методы класса
    public virtual void MakeSound()
    {
        Console.WriteLine($"{_name} издает звук...");
    }

    public virtual void Eat()
    {
        Console.WriteLine($"{_name} ест...");
    }

    public void Sleep()
    {
        Console.WriteLine($"{_name} спит...");
    }

    // Статический метод (принадлежит классу)
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Всего животных создано: {TotalAnimals}");
    }

    // Метод для отображения информации
    public override string ToString()
    {
        return $"Имя: {_name}, Возраст: {_age}, Вес: {_weight} кг";
    }
}

// Производный класс Dog
public class Dog : Animal
{
    // Дополнительные атрибуты
    private string _breed;
    private bool _isTrained;

    // Свойства
    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    public bool IsTrained
    {
        get { return _isTrained; }
        set { _isTrained = value; }
    }

    // Конструктор
    public Dog(string name, int age, double weight, string breed) 
        : base(name, age, weight)
    {
        Breed = breed;
        IsTrained = false;
    }

    // Переопределение метода базового класса
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} ест собачий корм...");
    }

    // Специфичный метод для Dog
    public void Fetch()
    {
        if (_isTrained)
            Console.WriteLine($"{Name} приносит мячик!");
        else
            Console.WriteLine($"{Name} не обучен приносить предметы.");
    }

    public void Train()
    {
        _isTrained = true;
        Console.WriteLine($"{Name} теперь обучен!");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    // Дополнительные атрибуты
    private string _color;
    private bool _isIndoor;

    // Свойства
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    public bool IsIndoor
    {
        get { return _isIndoor; }
        set { _isIndoor = value; }
    }

    // Конструктор
    public Cat(string name, int age, double weight, string color, bool isIndoor) 
        : base(name, age, weight)
    {
        Color = color;
        IsIndoor = isIndoor;
    }

    // Переопределение методов
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяяяу!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} ест кошачий корм...");
    }

    // Специфичные методы для Cat
    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет...");
    }

    public void Scratch()
    {
        Console.WriteLine($"{Name} точит когти...");
    }

    public void CheckLifestyle()
    {
        if (_isIndoor)
            Console.WriteLine($"{Name} - домашняя кошка");
        else
            Console.WriteLine($"{Name} - уличная кошка");
    }
}

// Производный класс Bird
public class Bird : Animal
{
    // Дополнительные атрибуты
    private double _wingSpan;
    private bool _canFly;

    // Свойства
    public double WingSpan
    {
        get { return _wingSpan; }
        set { _wingSpan = value; }
    }

    public bool CanFly
    {
        get { return _canFly; }
        set { _canFly = value; }
    }

    // Конструктор
    public Bird(string name, int age, double weight, double wingSpan) 
        : base(name, age, weight)
    {
        WingSpan = wingSpan;
        CanFly = true;
    }

    // Переопределение методов
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} поет: Чик-чирик!");
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} клюет зерна...");
    }

    // Специфичные методы для Bird
    public void Fly()
    {
        if (_canFly)
            Console.WriteLine($"{Name} летит на высоте...");
        else
            Console.WriteLine($"{Name} не может летать!");
    }

    public void BuildNest()
    {
        Console.WriteLine($"{Name} строит гнездо...");
    }
}

// Класс Program для демонстрации
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== Создание животных ===\n");

        // Создание объектов
        Dog dog = new Dog("Шарик", 3, 15.5, "Немецкая овчарка");
        Cat cat = new Cat("Мурзик", 2, 4.2, "Рыжий", true);
        Bird bird = new Bird("Кеша", 1, 0.5, 30.0);

        // Демонстрация свойств с геттерами и сеттерами
        Console.WriteLine("=== Использование свойств ===");
        Console.WriteLine($"Собака: {dog.Name}, Порода: {dog.Breed}");
        dog.Age = 4; // Использование сеттера
        Console.WriteLine($"Возраст собаки изменен на: {dog.Age}\n");

        // Демонстрация методов
        Console.WriteLine("=== Вызов методов ===");
        dog.MakeSound();
        dog.Eat();
        dog.Train();
        dog.Fetch();
        Console.WriteLine();

        cat.MakeSound();
        cat.Eat();
        cat.Purr();
        cat.Scratch();
        cat.CheckLifestyle();
        Console.WriteLine();

        bird.MakeSound();
        bird.Eat();
        bird.Fly();
        bird.BuildNest();
        Console.WriteLine();

        // Демонстрация статического метода
        Console.WriteLine("=== Статический метод ===");
        Animal.DisplayTotalAnimals();

        // Демонстрация переопределения методов
        Console.WriteLine("\n=== Полиморфизм ===");
        Animal[] animals = { dog, cat, bird };
        foreach (var animal in animals)
        {
            animal.MakeSound(); // Вызывается переопределенный метод
        }
    }
}
````
