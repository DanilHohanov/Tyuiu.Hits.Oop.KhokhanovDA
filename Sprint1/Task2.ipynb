<h1 style="color:DodgerBlue">Создание экземпляров классов (объектов) и взаимодействие объектов между собой</h1>

### Создание экземпляров классов (объектов)

В C#, экземпляр класса создается с помощью оператора `new`, который выделяет память для нового объекта и вызывает конструктор класса для инициализации этого объекта. Конструктор — это специальный метод в классе, который имеет то же имя, что и класс, и используется для установки начального состояния объекта.

#### Пример создания экземпляров классов Vehicle и Car

```csharp
Vehicle myVehicle = new Vehicle();
Car myCar = new Car();
```

В этом примере создаются два объекта: один для класса `Vehicle` и другой для класса `Car`. Конструкторы этих классов будут автоматически вызваны при создании объектов.

### Взаимодействие объектов между Собой

Взаимодействие объектов между собой означает, что один объект может вызывать методы другого объекта или обращаться к его свойствам (если они доступны). Это позволяет объектам обмениваться информацией и координировать свои действия.

#### Пример взаимодействия объектов

```csharp
myVehicle.Color = "Red";
myVehicle.Speed = 60;

myCar.Color = "Blue";
myCar.FuelLevel = 100;
myCar.Refuel(50); // Увеличиваем уровень топлива на 50
```

В этом примере мы устанавливаем свойства `Color` и `Speed` для объекта `myVehicle`, а также свойства `Color` и `FuelLevel` для объекта `myCar`. Затем мы вызываем метод `Refuel` у объекта `myCar`, чтобы увеличить его уровень топлива.

### Закрепление правил использования модификаторов доступа

Модификаторы доступа определяют уровень доступности членов класса из других частей кода. Важно правильно использовать модификаторы доступа для обеспечения безопасности и инкапсуляции данных.

#### Примеры модификаторов доступа в классах Vehicle и Car

```csharp
public class Vehicle
{
    public string Color { get; set; } // public свойство
    private int speed; // private поле

    protected void SetSpeed(int value) // protected метод
    {
        speed = value;
    }
}

public class Car : Vehicle
{
    private int fuelLevel; // private поле

    public void Refuel(int amount) // public метод
    {
        fuelLevel += amount;
    }

    protected void CheckSpeedLimit() // protected метод
    {
        if (speed > 120)
        {
            Console.WriteLine("Превышен скоростной режим!");
        }
    }
}
```

В этом примере:
- Свойство `Color` в классе `Vehicle` объявлено как `public`, что позволяет доступ к нему из любого места в коде.
- Поле `speed` в классе `Vehicle` объявлено как `private`, что ограничивает доступ к нему только внутри класса `Vehicle`.
- Метод `SetSpeed` в классе `Vehicle` объявлен как `protected`, что позволяет доступ к нему из класса `Vehicle` и его производных классов.
- Поле `fuelLevel` в классе `Car` объявлено как `private`, ограничивая доступ к нему только внутри класса `Car`.
- Методы `Refuel` и `CheckSpeedLimit` в классе `Car` демонстрируют использование модификаторов доступа для контроля над тем, как объекты могут взаимодействовать друг с другом.

### Заключение

Создание экземпляров классов и взаимодействие между объектами являются ключевыми аспектами объектно-ориентированного программирования в C#. Правильное использование модификаторов доступа обеспечивает безопасность данных и инкапсуляцию, позволяя разработчикам создавать более надежные и управляемые программы.

````
using System;

// Базовый класс Vehicle
public class Vehicle
{
    // Public свойство - доступно из любого места
    public string Color { get; set; }
    
    // Public свойство для скорости
    public int Speed { get; set; }
    
    // Private поле - доступно только внутри класса Vehicle
    private string _model;
    
    // Protected поле - доступно в Vehicle и производных классах
    protected int maxSpeed;

    // Конструктор
    public Vehicle(string model)
    {
        _model = model;
        Color = "White";
        Speed = 0;
        maxSpeed = 120;
    }

    // Public метод
    public void DisplayInfo()
    {
        Console.WriteLine($"Модель: {_model}, Цвет: {Color}, Скорость: {Speed} км/ч");
    }

    // Protected метод - доступен только в Vehicle и наследниках
    protected void SetMaxSpeed(int value)
    {
        maxSpeed = value;
    }

    protected int GetMaxSpeed()
    {
        return maxSpeed;
    }
}

// Производный класс Car
public class Car : Vehicle
{
    // Private поле - доступно только внутри класса Car
    private int fuelLevel;
    
    // Private поле
    private string _owner;

    // Public свойство
    public string Owner
    {
        get { return _owner; }
        set { _owner = value; }
    }

    // Конструктор
    public Car(string model, string owner) : base(model)
    {
        _owner = owner;
        fuelLevel = 0;
        maxSpeed = 180; // У машин максимальная скорость выше
    }

    // Public метод - заправка
    public void Refuel(int amount)
    {
        if (amount > 0)
        {
            fuelLevel += amount;
            Console.WriteLine($"{_owner} заправил {amount} литров. Текущий уровень топлива: {fuelLevel}л");
        }
    }

    // Public метод - проверка топлива
    public int GetFuelLevel()
    {
        return fuelLevel;
    }

    // Protected метод - проверка ограничения скорости
    protected void CheckSpeedLimit()
    {
        if (Speed > maxSpeed)
        {
            Console.WriteLine($"ВНИМАНИЕ: Превышена максимальная скорость ({maxSpeed} км/ч)!");
        }
        else
        {
            Console.WriteLine($"Скорость в пределах нормы (макс: {maxSpeed} км/ч)");
        }
    }

    // Public метод для ускорения (использует protected метод)
    public void Accelerate(int speed)
    {
        Speed += speed;
        Console.WriteLine($"{_owner} ускоряется. Текущая скорость: {Speed} км/ч");
        CheckSpeedLimit(); // Вызов protected метода
    }

    // Переопределение метода отображения информации
    public void DisplayCarInfo()
    {
        DisplayInfo(); // Вызов метода базового класса
        Console.WriteLine($"Владелец: {_owner}, Топливо: {fuelLevel}л");
    }
}

// Класс Program для демонстрации
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== Создание экземпляров классов ===\n");

        // Создание объектов с помощью оператора new
        Vehicle myVehicle = new Vehicle("Generic Vehicle");
        Car myCar = new Car("Toyota Camry", "Алексей");
        Car anotherCar = new Car("BMW X5", "Мария");

        Console.WriteLine("=== Взаимодействие с объектами ===\n");

        // Работа с Vehicle
        myVehicle.Color = "Red";
        myVehicle.Speed = 60;
        myVehicle.DisplayInfo();

        Console.WriteLine();

        // Работа с Car
        myCar.Color = "Blue";
        myCar.Speed = 0;
        myCar.Refuel(50); // Увеличиваем уровень топлива на 50
        myCar.DisplayCarInfo();

        Console.WriteLine();

        // Взаимодействие объектов между собой
        Console.WriteLine("=== Взаимодействие объектов ===\n");

        // Объект myCar взаимодействует со своими методами
        myCar.Accelerate(30);
        myCar.Accelerate(50);
        myCar.Accelerate(80); // Должно вызвать предупреждение о превышении

        Console.WriteLine();

        // Другой объект Car
        anotherCar.Color = "Black";
        anotherCar.Refuel(40);
        anotherCar.Accelerate(100);

        Console.WriteLine();

        // Доступ к public свойствам
        Console.WriteLine("=== Доступ к свойствам ===\n");
        Console.WriteLine($"Владелец myCar: {myCar.Owner}");
        Console.WriteLine($"Цвет myCar: {myCar.Color}");
        Console.WriteLine($"Уровень топлива: {myCar.GetFuelLevel()}л");

        Console.WriteLine();

        // Попытка доступа к private и protected членам (закомментирована, так как вызовет ошибку)
        // myCar.fuelLevel = 100; // ОШИБКА: fuelLevel - private, недоступен
        // myCar.maxSpeed = 200;  // ОШИБКА: maxSpeed - protected, недоступен извне класса
        // myCar.CheckSpeedLimit(); // ОШИБКА: CheckSpeedLimit() - protected, недоступен извне

        Console.WriteLine("=== Пример инкапсуляции ===\n");
        Console.WriteLine("Private поле 'fuelLevel' недоступно напрямую извне класса Car");
        Console.WriteLine("Protected поле 'maxSpeed' доступно только внутри Car и Vehicle");
        Console.WriteLine("Public свойства и методы доступны из любого места");
    }
}
````

````
using System;
using System.Collections.Generic;

// ═══════════════════════════════════════════════════════
// БАЗОВЫЙ КЛАСС ANIMAL
// ═══════════════════════════════════════════════════════
public class Animal
{
    // Статические атрибуты
    public static int TotalAnimals { get; private set; }
    protected static List<Animal> allAnimals = new List<Animal>();

    // Приватные поля
    private string _name;
    private int _age;
    private double _weight;
    private bool _isHungry;

    // Свойства с валидацией
    public string Name
    {
        get { return _name; }
        set { if (!string.IsNullOrEmpty(value) && value.Length <= 50) _name = value; }
    }

    public int Age
    {
        get { return _age; }
        set { if (value >= 0 && value <= 100) _age = value; }
    }

    public double Weight
    {
        get { return _weight; }
        set { if (value > 0 && value <= 1000) _weight = value; }
    }

    public bool IsHungry
    {
        get { return _isHungry; }
        set { _isHungry = value; }
    }

    // Конструктор
    public Animal(string name, int age, double weight)
    {
        Name = name;
        Age = age;
        Weight = weight;
        _isHungry = true;
        TotalAnimals++;
        allAnimals.Add(this);
    }

    // Виртуальные методы
    public virtual void MakeSound() => Console.WriteLine($"{_name} издает звук...");
    
    public virtual void Eat()
    {
        if (_isHungry) { Console.WriteLine($"{_name} ест..."); _isHungry = false; }
        else Console.WriteLine($"{_name} не голоден.");
    }

    public void Sleep() => Console.WriteLine($"{_name} спит...");
    public void Play() { Console.WriteLine($"{_name} играет..."); _isHungry = true; }

    // Статические методы
    public static void DisplayTotalAnimals() => Console.WriteLine($"✓ Всего животных: {TotalAnimals}");
    public static List<Animal> GetAllAnimals() => allAnimals;

    public override string ToString() => 
        $"[{_name}] Возраст: {_age}, Вес: {_weight}кг, Голоден: {(_isHungry ? "✓" : "✗")}";
}

// ═══════════════════════════════════════════════════════
// ПРОИЗВОДНЫЙ КЛАСС DOG
// ═══════════════════════════════════════════════════════
public class Dog : Animal
{
    private string _breed;
    private bool _isTrained;
    private int _loyaltyLevel;

    public string Breed { get => _breed; set => _breed = value; }
    public bool IsTrained { get => _isTrained; set => _isTrained = value; }
    public int LoyaltyLevel => _loyaltyLevel;

    public Dog(string name, int age, double weight, string breed) : base(name, age, weight)
    {
        Breed = breed;
        IsTrained = false;
        _loyaltyLevel = 50;
    }

    public override void MakeSound() => Console.WriteLine($"{Name} лает: Гав-гав! 🐕");
    
    public override void Eat()
    {
        if (IsHungry) { Console.WriteLine($"{Name} ест собачий корм... 🦴"); IsHungry = false; _loyaltyLevel = Math.Min(100, _loyaltyLevel + 10); }
        else Console.WriteLine($"{Name} не голоден.");
    }

    public void Bark() => Console.WriteLine($"{Name} громко лает: ГАВ-ГАВ! 🔊");
    
    public void Fetch()
    {
        if (_isTrained) Console.WriteLine($"{Name} приносит мячик! 🎾 Лояльность: {_loyaltyLevel}/100");
        else Console.WriteLine($"{Name} не обучен приносить предметы.");
    }

    public void Train() { _isTrained = true; _loyaltyLevel = Math.Min(100, _loyaltyLevel + 20); Console.WriteLine($"{Name} обучен! 🎓 Лояльность: {_loyaltyLevel}/100"); }
    
    public void InteractWithAnimal(Animal other) { Console.WriteLine($"{Name} 🤝 {other.Name}"); MakeSound(); }
    
    public void PlayWithDog(Dog other) { Console.WriteLine($"{Name} + {other.Name} = веселье! 🎉"); Play(); other.Play(); }
}

// ═══════════════════════════════════════════════════════
// ПРОИЗВОДНЫЙ КЛАСС CAT
// ═══════════════════════════════════════════════════════
public class Cat : Animal
{
    private string _color;
    private bool _isIndoor;
    private int _moodLevel;

    public string Color { get => _color; set => _color = value; }
    public bool IsIndoor { get => _isIndoor; set => _isIndoor = value; }
    public int MoodLevel => _moodLevel;

    public Cat(string name, int age, double weight, string color, bool isIndoor) : base(name, age, weight)
    {
        Color = color;
        IsIndoor = isIndoor;
        _moodLevel = 50;
    }

    public override void MakeSound() => Console.WriteLine($"{Name} мяукает: Мяяяу! 🐱");
    
    public override void Eat()
    {
        if (IsHungry) { Console.WriteLine($"{Name} ест кошачий корм... 🐟"); IsHungry = false; _moodLevel = Math.Min(100, _moodLevel + 15); }
        else Console.WriteLine($"{Name} не голоден.");
    }

    public void Meow() => Console.WriteLine($"{Name} тихо: мяу... 🤫");
    
    public void Purr()
    {
        if (_moodLevel > 30) Console.WriteLine($"{Name} мурлычет... 😊 Настроение: {_moodLevel}/100");
        else Console.WriteLine($"{Name} не в настроении 😾");
    }

    public void Scratch() { Console.WriteLine($"{Name} точит когти... 🪵"); _moodLevel = Math.Max(0, _moodLevel - 5); }
    
    public void CheckLifestyle() => Console.WriteLine($"{Name} - {(_isIndoor ? "🏠 домашняя" : "🌳 уличная")} кошка");
    
    public void InteractWithDog(Dog dog)
    {
        Console.WriteLine($"{Name} видит {dog.Name}... 👀");
        if (dog.LoyaltyLevel > 70) { Console.WriteLine($"{Name} спокойно наблюдает 😌"); MakeSound(); }
        else { Console.WriteLine($"{Name} шипит! 😾"); }
    }
    
    public void PlayWithCat(Cat other) { Console.WriteLine($"{Name} + {other.Name} = игры! 🧶"); _moodLevel = Math.Min(100, _moodLevel + 20); }
}

// ═══════════════════════════════════════════════════════
// ПРОИЗВОДНЫЙ КЛАСС BIRD
// ═══════════════════════════════════════════════════════
public class Bird : Animal
{
    private double _wingSpan;
    private bool _canFly;
    private int _altitude;

    public double WingSpan { get => _wingSpan; set => _wingSpan = value; }
    public bool CanFly { get => _canFly; set => _canFly = value; }
    public int Altitude => _altitude;

    public Bird(string name, int age, double weight, double wingSpan) : base(name, age, weight)
    {
        WingSpan = wingSpan;
        CanFly = true;
        _altitude = 0;
    }

    public override void MakeSound() => Console.WriteLine($"{Name} поет: Чик-чирик! 🐦");
    
    public override void Eat()
    {
        if (IsHungry) { Console.WriteLine($"{Name} клюет зерна... 🌾"); IsHungry = false; }
        else Console.WriteLine($"{Name} не голоден.");
    }

    public void Fly(int height) { if (_canFly) { _altitude = height; Console.WriteLine($"{Name} летит на высоте {height}м! ✈️"); } else Console.WriteLine($"{Name} не может летать!"); }
    public void Sing() => Console.WriteLine($"{Name} поёт песню! 🎵");
    public void BuildNest() => Console.WriteLine($"{Name} строит гнездо... 🪺");
    public void FlyAroundAnimal(Animal animal) { Console.WriteLine($"{Name} кружит вокруг {animal.Name}... 🌀"); Fly(10); }
}

// ═══════════════════════════════════════════════════════
// 🎬 ДЕМОСТРАЦИЯ РАБОТЫ КЛАССОВ
// ═══════════════════════════════════════════════════════

Console.WriteLine("╔════════════════════════════════════════╗");
Console.WriteLine("║  🐾 ДЕМОСТРАЦИЯ КЛАССОВ ЖИВОТНЫХ 🐾   ║");
Console.WriteLine("╚════════════════════════════════════════╝\n");

// ─────────────────────────────────────────────────────────
// 1️⃣ СОЗДАНИЕ ОБЪЕКТОВ
// ─────────────────────────────────────────────────────────
Console.WriteLine("📦 1. СОЗДАНИЕ ЭКЗЕМПЛЯРОВ:");
Console.WriteLine("────────────────────────────");
Dog dog1 = new Dog("Шарик", 3, 15.5, "Немецкая овчарка");
Dog dog2 = new Dog("Бобик", 2, 12.0, "Лабрадор");
Cat cat1 = new Cat("Мурзик", 2, 4.2, "Рыжий", true);
Cat cat2 = new Cat("Барсик", 1, 3.8, "Черный", false);
Bird bird1 = new Bird("Кеша", 1, 0.5, 30.0);
Console.WriteLine("✓ Создано: 2 собаки, 2 кошки, 1 птица ✓\n");

// ─────────────────────────────────────────────────────────
// 2️⃣ СВОЙСТВА: ГЕТТЕРЫ/СЕТТЕРЫ
// ─────────────────────────────────────────────────────────
Console.WriteLine("🔧 2. РАБОТА СО СВОЙСТВАМИ:");
Console.WriteLine("────────────────────────────");
Console.WriteLine($"🐕 {dog1.Name} | Порода: {dog1.Breed} | Возраст: {dog1.Age}");
dog1.Age = 4;
Console.WriteLine($"✓ Возраст {dog1.Name} изменён: {dog1.Age}\n");

Console.WriteLine($"🐱 {cat1.Name} | Цвет: {cat1.Color}");
cat1.Color = "Бело-рыжий";
Console.WriteLine($"✓ Цвет изменён: {cat1.Color}\n");

// Валидация
Console.WriteLine("🛡️ Проверка валидации:");
cat1.Age = 150; // Не изменится (>100)
Console.WriteLine($"Попытка установить возраст 150 → реальный: {cat1.Age} ✓\n");

// ─────────────────────────────────────────────────────────
// 3️⃣ МЕТОДЫ БАЗОВОГО КЛАССА
// ─────────────────────────────────────────────────────────
Console.WriteLine("🎯 3. МЕТОДЫ БАЗОВОГО КЛАССA:");
Console.WriteLine("─────────────────────────────");
Animal generic = new Animal("Зверь", 5, 10.0);
generic.MakeSound();
generic.Eat();
generic.Sleep();
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 4️⃣ ПОЛИМОРФИЗМ: ПЕРЕОПРЕДЕЛЕНИЕ МЕТОДОВ
// ─────────────────────────────────────────────────────────
Console.WriteLine("🔄 4. ПОЛИМОРФИЗМ:");
Console.WriteLine("──────────────────");
Console.WriteLine("🐕 Собака:");
dog1.MakeSound();
dog1.Eat();
Console.WriteLine();

Console.WriteLine("🐱 Кошка:");
cat1.MakeSound();
cat1.Eat();
Console.WriteLine();

Console.WriteLine("🐦 Птица:");
bird1.MakeSound();
bird1.Eat();
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 5️⃣ СПЕЦИФИЧНЫЕ МЕТОДЫ ПРОИЗВОДНЫХ КЛАССОВ
// ─────────────────────────────────────────────────────────
Console.WriteLine("⭐ 5. СПЕЦИФИЧНЫЕ МЕТОДЫ:");
Console.WriteLine("─────────────────────────");

Console.WriteLine("🐕 Собака:");
dog1.Train();
dog1.Fetch();
dog1.Bark();
Console.WriteLine();

Console.WriteLine("🐱 Кошка:");
cat1.Purr();
cat1.Scratch();
cat1.CheckLifestyle();
Console.WriteLine();

Console.WriteLine("🐦 Птица:");
bird1.Fly(150);
bird1.Sing();
bird1.BuildNest();
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 6️⃣ ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ
// ─────────────────────────────────────────────────────────
Console.WriteLine("🤝 6. ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ:");
Console.WriteLine("──────────────────────────────");

Console.WriteLine("→ Собаки играют:");
dog1.PlayWithDog(dog2);
Console.WriteLine();

Console.WriteLine("→ Кошки играют:");
cat1.PlayWithCat(cat2);
Console.WriteLine();

Console.WriteLine("→ Собака и кошка:");
dog1.InteractWithAnimal(cat1);
cat1.InteractWithDog(dog1);
Console.WriteLine();

Console.WriteLine("→ Птица кружит:");
bird1.FlyAroundAnimal(dog1);
bird1.FlyAroundAnimal(cat1);
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 7️⃣ СОВМЕСТНАЯ АКТИВНОСТЬ
// ─────────────────────────────────────────────────────────
Console.WriteLine("🎮 7. СОВМЕСТНАЯ АКТИВНОСТЬ:");
Console.WriteLine("────────────────────────────");
Console.WriteLine("→ Все играют:");
dog1.Play();
cat1.Play();
bird1.Play();
Console.WriteLine();

Console.WriteLine("→ Кормление после игры:");
dog1.Eat();
cat1.Eat();
bird1.Eat();
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 8️⃣ СТАТИЧЕСКИЕ ЧЛЕНЫ
// ─────────────────────────────────────────────────────────
Console.WriteLine("📊 8. СТАТИЧЕСКИЕ ЧЛЕНЫ:");
Console.WriteLine("────────────────────────");
Animal.DisplayTotalAnimals();
List<Animal> all = Animal.GetAllAnimals();
Console.WriteLine($"✓ В реестре: {all.Count} животных ✓\n");

// ─────────────────────────────────────────────────────────
// 9️⃣ ПОЛИМОРФИЗМ ЧЕРЕЗ МАССИВ
// ─────────────────────────────────────────────────────────
Console.WriteLine("🎭 9. ПОЛИМОРФИЗМ (массив базового типа):");
Console.WriteLine("────────────────────────────────────────");
Animal[] animals = { dog1, dog2, cat1, cat2, bird1 };
Console.WriteLine("Вызов MakeSound() для всех:");
foreach (var a in animals) a.MakeSound();
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 🔟 СОСТОЯНИЯ ОБЪЕКТОВ (ToString)
// ─────────────────────────────────────────────────────────
Console.WriteLine("📋 10. СОСТОЯНИЯ ОБЪЕКТОВ:");
Console.WriteLine("──────────────────────────");
Console.WriteLine(dog1);
Console.WriteLine(cat1);
Console.WriteLine(bird1);
Console.WriteLine();

// ─────────────────────────────────────────────────────────
// 🏁 ИТОГИ
// ─────────────────────────────────────────────────────────
Console.WriteLine("╔════════════════════════════════════════╗");
Console.WriteLine("║           🏁 ИТОГИ 🏁                  ║");
Console.WriteLine("╚════════════════════════════════════════╝");
Console.WriteLine($"✓ Всего создано: {Animal.TotalAnimals} животных");
Console.WriteLine("✓ Продемонстрировано:");
Console.WriteLine("  • Наследование (Dog, Cat, Bird → Animal)");
Console.WriteLine("  • Инкапсуляция (приватные поля + свойства)");
Console.WriteLine("  • Полиморфизм (переопределение методов)");
Console.WriteLine("  • Статические члены класса");
Console.WriteLine("  • Валидация в сеттерах");
Console.WriteLine("  • Взаимодействие объектов");
Console.WriteLine("  • Специфичные методы производных классов");
Console.WriteLine("\n🎉 Демонстрация завершена успешно! 🎉");
````
