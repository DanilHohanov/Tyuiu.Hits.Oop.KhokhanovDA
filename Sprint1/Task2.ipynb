<h1 style="color:DodgerBlue">Создание экземпляров классов (объектов) и взаимодействие объектов между собой</h1>

### Создание экземпляров классов (объектов)

В C#, экземпляр класса создается с помощью оператора `new`, который выделяет память для нового объекта и вызывает конструктор класса для инициализации этого объекта. Конструктор — это специальный метод в классе, который имеет то же имя, что и класс, и используется для установки начального состояния объекта.

#### Пример создания экземпляров классов Vehicle и Car

```csharp
Vehicle myVehicle = new Vehicle();
Car myCar = new Car();
```

В этом примере создаются два объекта: один для класса `Vehicle` и другой для класса `Car`. Конструкторы этих классов будут автоматически вызваны при создании объектов.

### Взаимодействие объектов между Собой

Взаимодействие объектов между собой означает, что один объект может вызывать методы другого объекта или обращаться к его свойствам (если они доступны). Это позволяет объектам обмениваться информацией и координировать свои действия.

#### Пример взаимодействия объектов

```csharp
myVehicle.Color = "Red";
myVehicle.Speed = 60;

myCar.Color = "Blue";
myCar.FuelLevel = 100;
myCar.Refuel(50); // Увеличиваем уровень топлива на 50
```

В этом примере мы устанавливаем свойства `Color` и `Speed` для объекта `myVehicle`, а также свойства `Color` и `FuelLevel` для объекта `myCar`. Затем мы вызываем метод `Refuel` у объекта `myCar`, чтобы увеличить его уровень топлива.

### Закрепление правил использования модификаторов доступа

Модификаторы доступа определяют уровень доступности членов класса из других частей кода. Важно правильно использовать модификаторы доступа для обеспечения безопасности и инкапсуляции данных.

#### Примеры модификаторов доступа в классах Vehicle и Car

```csharp
public class Vehicle
{
    public string Color { get; set; } // public свойство
    private int speed; // private поле

    protected void SetSpeed(int value) // protected метод
    {
        speed = value;
    }
}

public class Car : Vehicle
{
    private int fuelLevel; // private поле

    public void Refuel(int amount) // public метод
    {
        fuelLevel += amount;
    }

    protected void CheckSpeedLimit() // protected метод
    {
        if (speed > 120)
        {
            Console.WriteLine("Превышен скоростной режим!");
        }
    }
}
```

В этом примере:
- Свойство `Color` в классе `Vehicle` объявлено как `public`, что позволяет доступ к нему из любого места в коде.
- Поле `speed` в классе `Vehicle` объявлено как `private`, что ограничивает доступ к нему только внутри класса `Vehicle`.
- Метод `SetSpeed` в классе `Vehicle` объявлен как `protected`, что позволяет доступ к нему из класса `Vehicle` и его производных классов.
- Поле `fuelLevel` в классе `Car` объявлено как `private`, ограничивая доступ к нему только внутри класса `Car`.
- Методы `Refuel` и `CheckSpeedLimit` в классе `Car` демонстрируют использование модификаторов доступа для контроля над тем, как объекты могут взаимодействовать друг с другом.

### Заключение

Создание экземпляров классов и взаимодействие между объектами являются ключевыми аспектами объектно-ориентированного программирования в C#. Правильное использование модификаторов доступа обеспечивает безопасность данных и инкапсуляцию, позволяя разработчикам создавать более надежные и управляемые программы.

````
using System;

// Базовый класс Vehicle
public class Vehicle
{
    // Public свойство - доступно из любого места
    public string Color { get; set; }
    
    // Public свойство для скорости
    public int Speed { get; set; }
    
    // Private поле - доступно только внутри класса Vehicle
    private string _model;
    
    // Protected поле - доступно в Vehicle и производных классах
    protected int maxSpeed;

    // Конструктор
    public Vehicle(string model)
    {
        _model = model;
        Color = "White";
        Speed = 0;
        maxSpeed = 120;
    }

    // Public метод
    public void DisplayInfo()
    {
        Console.WriteLine($"Модель: {_model}, Цвет: {Color}, Скорость: {Speed} км/ч");
    }

    // Protected метод - доступен только в Vehicle и наследниках
    protected void SetMaxSpeed(int value)
    {
        maxSpeed = value;
    }

    protected int GetMaxSpeed()
    {
        return maxSpeed;
    }
}

// Производный класс Car
public class Car : Vehicle
{
    // Private поле - доступно только внутри класса Car
    private int fuelLevel;
    
    // Private поле
    private string _owner;

    // Public свойство
    public string Owner
    {
        get { return _owner; }
        set { _owner = value; }
    }

    // Конструктор
    public Car(string model, string owner) : base(model)
    {
        _owner = owner;
        fuelLevel = 0;
        maxSpeed = 180; // У машин максимальная скорость выше
    }

    // Public метод - заправка
    public void Refuel(int amount)
    {
        if (amount > 0)
        {
            fuelLevel += amount;
            Console.WriteLine($"{_owner} заправил {amount} литров. Текущий уровень топлива: {fuelLevel}л");
        }
    }

    // Public метод - проверка топлива
    public int GetFuelLevel()
    {
        return fuelLevel;
    }

    // Protected метод - проверка ограничения скорости
    protected void CheckSpeedLimit()
    {
        if (Speed > maxSpeed)
        {
            Console.WriteLine($"ВНИМАНИЕ: Превышена максимальная скорость ({maxSpeed} км/ч)!");
        }
        else
        {
            Console.WriteLine($"Скорость в пределах нормы (макс: {maxSpeed} км/ч)");
        }
    }

    // Public метод для ускорения (использует protected метод)
    public void Accelerate(int speed)
    {
        Speed += speed;
        Console.WriteLine($"{_owner} ускоряется. Текущая скорость: {Speed} км/ч");
        CheckSpeedLimit(); // Вызов protected метода
    }

    // Переопределение метода отображения информации
    public void DisplayCarInfo()
    {
        DisplayInfo(); // Вызов метода базового класса
        Console.WriteLine($"Владелец: {_owner}, Топливо: {fuelLevel}л");
    }
}

// Класс Program для демонстрации
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== Создание экземпляров классов ===\n");

        // Создание объектов с помощью оператора new
        Vehicle myVehicle = new Vehicle("Generic Vehicle");
        Car myCar = new Car("Toyota Camry", "Алексей");
        Car anotherCar = new Car("BMW X5", "Мария");

        Console.WriteLine("=== Взаимодействие с объектами ===\n");

        // Работа с Vehicle
        myVehicle.Color = "Red";
        myVehicle.Speed = 60;
        myVehicle.DisplayInfo();

        Console.WriteLine();

        // Работа с Car
        myCar.Color = "Blue";
        myCar.Speed = 0;
        myCar.Refuel(50); // Увеличиваем уровень топлива на 50
        myCar.DisplayCarInfo();

        Console.WriteLine();

        // Взаимодействие объектов между собой
        Console.WriteLine("=== Взаимодействие объектов ===\n");

        // Объект myCar взаимодействует со своими методами
        myCar.Accelerate(30);
        myCar.Accelerate(50);
        myCar.Accelerate(80); // Должно вызвать предупреждение о превышении

        Console.WriteLine();

        // Другой объект Car
        anotherCar.Color = "Black";
        anotherCar.Refuel(40);
        anotherCar.Accelerate(100);

        Console.WriteLine();

        // Доступ к public свойствам
        Console.WriteLine("=== Доступ к свойствам ===\n");
        Console.WriteLine($"Владелец myCar: {myCar.Owner}");
        Console.WriteLine($"Цвет myCar: {myCar.Color}");
        Console.WriteLine($"Уровень топлива: {myCar.GetFuelLevel()}л");

        Console.WriteLine();

        // Попытка доступа к private и protected членам (закомментирована, так как вызовет ошибку)
        // myCar.fuelLevel = 100; // ОШИБКА: fuelLevel - private, недоступен
        // myCar.maxSpeed = 200;  // ОШИБКА: maxSpeed - protected, недоступен извне класса
        // myCar.CheckSpeedLimit(); // ОШИБКА: CheckSpeedLimit() - protected, недоступен извне

        Console.WriteLine("=== Пример инкапсуляции ===\n");
        Console.WriteLine("Private поле 'fuelLevel' недоступно напрямую извне класса Car");
        Console.WriteLine("Protected поле 'maxSpeed' доступно только внутри Car и Vehicle");
        Console.WriteLine("Public свойства и методы доступны из любого места");
    }
}
````

Ниже в блоке по примеру создайте базовый класс Animal и производные классы (2-3 например Dog, Cat и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также статические и обычными атрибуты и метода класса с реализацией взаимодействия между объектами.

````
using System;
using System.Collections.Generic;

// Базовый класс Animal
public class Animal
{
    // Статический атрибут (общий для всех экземпляров)
    public static int TotalAnimals { get; private set; }
    
    // Статическое поле для хранения всех созданных животных
    protected static List<Animal> allAnimals = new List<Animal>();

    // Обычные атрибуты (private поля)
    private string _name;
    private int _age;
    private double _weight;
    private bool _isHungry;

    // Свойства с геттерами и сеттерами
    public string Name
    {
        get { return _name; }
        set 
        { 
            if (!string.IsNullOrEmpty(value) && value.Length <= 50)
                _name = value;
        }
    }

    public int Age
    {
        get { return _age; }
        set 
        { 
            if (value >= 0 && value <= 100)
                _age = value;
        }
    }

    public double Weight
    {
        get { return _weight; }
        set 
        { 
            if (value > 0 && value <= 1000)
                _weight = value;
        }
    }

    public bool IsHungry
    {
        get { return _isHungry; }
        set { _isHungry = value; }
    }

    // Конструктор
    public Animal(string name, int age, double weight)
    {
        Name = name;
        Age = age;
        Weight = weight;
        _isHungry = true;
        TotalAnimals++;
        allAnimals.Add(this);
    }

    // Обычные виртуальные методы
    public virtual void MakeSound()
    {
        Console.WriteLine($"{_name} издает звук...");
    }

    public virtual void Eat()
    {
        if (_isHungry)
        {
            Console.WriteLine($"{_name} ест...");
            _isHungry = false;
        }
        else
        {
            Console.WriteLine($"{_name} не голоден.");
        }
    }

    public void Sleep()
    {
        Console.WriteLine($"{_name} спит...");
    }

    public void Play()
    {
        Console.WriteLine($"{_name} играет...");
        _isHungry = true;
    }

    // Статический метод
    public static void DisplayTotalAnimals()
    {
        Console.WriteLine($"Всего животных создано: {TotalAnimals}");
    }

    // Статический метод для получения всех животных
    public static List<Animal> GetAllAnimals()
    {
        return allAnimals;
    }

    // Метод для отображения информации
    public override string ToString()
    {
        return $"Имя: {_name}, Возраст: {_age}, Вес: {_weight} кг, Голоден: {(_isHungry ? "да" : "нет")}";
    }
}

// Производный класс Dog
public class Dog : Animal
{
    // Дополнительные private атрибуты
    private string _breed;
    private bool _isTrained;
    private int _loyaltyLevel;

    // Свойства
    public string Breed
    {
        get { return _breed; }
        set { _breed = value; }
    }

    public bool IsTrained
    {
        get { return _isTrained; }
        set { _isTrained = value; }
    }

    public int LoyaltyLevel
    {
        get { return _loyaltyLevel; }
    }

    // Конструктор
    public Dog(string name, int age, double weight, string breed) 
        : base(name, age, weight)
    {
        Breed = breed;
        IsTrained = false;
        _loyaltyLevel = 50;
    }

    // Переопределение методов
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public override void Eat()
    {
        if (IsHungry)
        {
            Console.WriteLine($"{Name} ест собачий корм...");
            IsHungry = false;
            _loyaltyLevel = Math.Min(100, _loyaltyLevel + 10);
        }
        else
        {
            Console.WriteLine($"{Name} не голоден.");
        }
    }

    // Специфичные методы
    public void Bark()
    {
        Console.WriteLine($"{Name} громко лает: ГАВ-ГАВ!");
    }

    public void Fetch()
    {
        if (_isTrained)
        {
            Console.WriteLine($"{Name} приносит мячик! Уровень лояльности: {_loyaltyLevel}");
        }
        else
        {
            Console.WriteLine($"{Name} не обучен приносить предметы.");
        }
    }

    public void Train()
    {
        _isTrained = true;
        _loyaltyLevel = Math.Min(100, _loyaltyLevel + 20);
        Console.WriteLine($"{Name} теперь обучен! Уровень лояльности: {_loyaltyLevel}");
    }

    // Взаимодействие с другими животными
    public void InteractWithAnimal(Animal otherAnimal)
    {
        Console.WriteLine($"{Name} взаимодействует с {otherAnimal.Name}...");
        MakeSound();
    }

    // Взаимодействие с другой собакой
    public void PlayWithDog(Dog otherDog)
    {
        Console.WriteLine($"{Name} и {otherDog.Name} играют вместе!");
        otherDog.Play();
        this.Play();
    }
}

// Производный класс Cat
public class Cat : Animal
{
    // Дополнительные private атрибуты
    private string _color;
    private bool _isIndoor;
    private int _moodLevel;

    // Свойства
    public string Color
    {
        get { return _color; }
        set { _color = value; }
    }

    public bool IsIndoor
    {
        get { return _isIndoor; }
        set { _isIndoor = value; }
    }

    public int MoodLevel
    {
        get { return _moodLevel; }
    }

    // Конструктор
    public Cat(string name, int age, double weight, string color, bool isIndoor) 
        : base(name, age, weight)
    {
        Color = color;
        IsIndoor = isIndoor;
        _moodLevel = 50;
    }

    // Переопределение методов
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяяяу!");
    }

    public override void Eat()
    {
        if (IsHungry)
        {
            Console.WriteLine($"{Name} ест кошачий корм...");
            IsHungry = false;
            _moodLevel = Math.Min(100, _moodLevel + 15);
        }
        else
        {
            Console.WriteLine($"{Name} не голоден.");
        }
    }

    // Специфичные методы
    public void Meow()
    {
        Console.WriteLine($"{Name} тихо мяукает: мяу...");
    }

    public void Purr()
    {
        if (_moodLevel > 30)
        {
            Console.WriteLine($"{Name} мурлычет от удовольствия... Уровень настроения: {_moodLevel}");
        }
        else
        {
            Console.WriteLine($"{Name} не в настроении мурлыкать.");
        }
    }

    public void Scratch()
    {
        Console.WriteLine($"{Name} точит когти о когтеточку...");
        _moodLevel = Math.Max(0, _moodLevel - 5);
    }

    public void CheckLifestyle()
    {
        if (_isIndoor)
            Console.WriteLine($"{Name} - домашняя кошка");
        else
            Console.WriteLine($"{Name} - уличная кошка");
    }

    // Взаимодействие с собакой
    public void InteractWithDog(Dog dog)
    {
        Console.WriteLine($"{Name} видит {dog.Name}...");
        if (dog.LoyaltyLevel > 70)
        {
            Console.WriteLine($"{Name} спокойно наблюдает за {dog.Name}.");
            MakeSound();
        }
        else
        {
            Console.WriteLine($"{Name} шипит на {dog.Name}!");
        }
    }

    // Взаимодействие с другой кошкой
    public void PlayWithCat(Cat otherCat)
    {
        Console.WriteLine($"{Name} и {otherCat.Name} играют вместе!");
        _moodLevel = Math.Min(100, _moodLevel + 20);
        otherCat._moodLevel = Math.Min(100, otherCat._moodLevel + 20);
    }
}

// Производный класс Bird
public class Bird : Animal
{
    // Дополнительные private атрибуты
    private double _wingSpan;
    private bool _canFly;
    private int _altitude;

    // Свойства
    public double WingSpan
    {
        get { return _wingSpan; }
        set { _wingSpan = value; }
    }

    public bool CanFly
    {
        get { return _canFly; }
        set { _canFly = value; }
    }

    public int Altitude
    {
        get { return _altitude; }
    }

    // Конструктор
    public Bird(string name, int age, double weight, double wingSpan) 
        : base(name, age, weight)
    {
        WingSpan = wingSpan;
        CanFly = true;
        _altitude = 0;
    }

    // Переопределение методов
    public override void MakeSound()
    {
        Console.WriteLine($"{Name} поет: Чик-чирик!");
    }

    public override void Eat()
    {
        if (IsHungry)
        {
            Console.WriteLine($"{Name} клюет зерна...");
            IsHungry = false;
        }
        else
        {
            Console.WriteLine($"{Name} не голоден.");
        }
    }

    // Специфичные методы
    public void Fly(int height)
    {
        if (_canFly)
        {
            _altitude = height;
            Console.WriteLine($"{Name} летит на высоте {height} метров!");
        }
        else
        {
            Console.WriteLine($"{Name} не может летать!");
        }
    }

    public void Sing()
    {
        Console.WriteLine($"{Name} поет песню!");
    }

    public void BuildNest()
    {
        Console.WriteLine($"{Name} строит гнездо...");
    }

    // Взаимодействие с другими животными
    public void FlyAroundAnimal(Animal animal)
    {
        Console.WriteLine($"{Name} летает вокруг {animal.Name}...");
        Fly(10);
    }
}

// Класс Program для демонстрации
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== Создание экземпляров классов ===\n");

        // Создание объектов с помощью оператора new
        Dog dog1 = new Dog("Шарик", 3, 15.5, "Немецкая овчарка");
        Dog dog2 = new Dog("Бобик", 2, 12.0, "Лабрадор");
        
        Cat cat1 = new Cat("Мурзик", 2, 4.2, "Рыжий", true);
        Cat cat2 = new Cat("Барсик", 1, 3.8, "Черный", false);
        
        Bird bird1 = new Bird("Кеша", 1, 0.5, 30.0);

        Console.WriteLine("=== Использование свойств с геттерами и сеттерами ===\n");
        
        // Использование геттеров и сеттеров
        Console.WriteLine($"Собака: {dog1.Name}, Порода: {dog1.Breed}");
        dog1.Age = 4; // Использование сеттера
        Console.WriteLine($"Возраст {dog1.Name} изменен на: {dog1.Age}\n");

        Console.WriteLine($"Кошка: {cat1.Name}, Цвет: {cat1.Color}");
        cat1.Color = "Бело-рыжий";
        Console.WriteLine($"Цвет изменен на: {cat1.Color}\n");

        Console.WriteLine("=== Вызов методов ===\n");

        // Демонстрация методов
        dog1.MakeSound();
        dog1.Eat();
        dog1.Train();
        dog1.Fetch();
        Console.WriteLine();

        cat1.MakeSound();
        cat1.Eat();
        cat1.Purr();
        cat1.Scratch();
        cat1.CheckLifestyle();
        Console.WriteLine();

        bird1.MakeSound();
        bird1.Eat();
        bird1.Fly(150);
        bird1.Sing();
        bird1.BuildNest();
        Console.WriteLine();

        Console.WriteLine("=== ВЗАИМОДЕЙСТВИЕ ОБЪЕКТОВ МЕЖДУ СОБОЙ ===\n");

        // Взаимодействие между собаками
        Console.WriteLine("--- Взаимодействие между собаками ---");
        dog1.PlayWithDog(dog2);
        Console.WriteLine();

        // Взаимодействие между кошками
        Console.WriteLine("--- Взаимодействие между кошками ---");
        cat1.PlayWithCat(cat2);
        Console.WriteLine();

        // Взаимодействие собаки и кошки
        Console.WriteLine("--- Взаимодействие собаки и кошки ---");
        dog1.InteractWithAnimal(cat1);
        cat1.InteractWithDog(dog1);
        Console.WriteLine();

        // Взаимодействие птицы с другими животными
        Console.WriteLine("--- Взаимодействие птицы с животными ---");
        bird1.FlyAroundAnimal(dog1);
        bird1.FlyAroundAnimal(cat1);
        Console.WriteLine();

        // Совместная игра всех животных
        Console.WriteLine("--- Совместная активность ---");
        dog1.Play();
        cat1.Play();
        bird1.Play();
        Console.WriteLine();

        // После игры животные проголодались
        Console.WriteLine("--- Кормление после игры ---");
        dog1.Eat();
        cat1.Eat();
        bird1.Eat();
        Console.WriteLine();

        Console.WriteLine("=== Статические методы и атрибуты ===\n");
        
        // Использование статического метода
        Animal.DisplayTotalAnimals();
        
        // Получение списка всех животных
        List<Animal> allAnimals = Animal.GetAllAnimals();
        Console.WriteLine($"В списке зарегистрировано: {allAnimals.Count} животных\n");

        Console.WriteLine("=== Полиморфизм ===\n");
        
        // Демонстрация полиморфизма
        Animal[] animals = { dog1, dog2, cat1, cat2, bird1 };
        foreach (var animal in animals)
        {
            animal.MakeSound(); // Вызывается переопределенный метод
        }
        
        Console.WriteLine();

        Console.WriteLine("=== Проверка состояний объектов ===\n");
        Console.WriteLine(dog1.ToString());
        Console.WriteLine(cat1.ToString());
        Console.WriteLine(bird1.ToString());
    }
}
````
