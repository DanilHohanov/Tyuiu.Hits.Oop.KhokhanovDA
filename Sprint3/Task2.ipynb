<h1 style="color:DodgerBlue">Generic классы в C#</h1>


Генерические классы в C# представляют собой шаблон класса, который может работать с различными типами данных без потери производительности. Они позволяют создавать компоненты, которые могут быть безопасно повторно использованы в различных контекстах, обеспечивая при этом строгую типизацию и избегая необходимости приведения типов.

#### Основные преимущества генерических классов:

1. **Безопасность типов**: Генерические классы обеспечивают проверку типов на этапе компиляции, что помогает предотвратить ошибки во время выполнения.
2. **Универсальность**: Генерические классы могут работать с различными типами данных, делая их очень гибкими и универсальными.
3. **Производительность**: Генерические классы не приводят к значительной потере производительности, поскольку информация о типе хранится вместе с объектом.

#### Пример с классами `Vehicle`, `Car` и `Motorcycle`

Давайте рассмотрим пример генерического класса `VehicleCollection<T>`, который может хранить коллекцию объектов типа `T`, где `T` может быть любой строкой, например, `Vehicle`, `Car` или `Motorcycle`.

```csharp
using System;
using System.Collections.Generic;

// Определение базового класса Vehicle
public class Vehicle
{
    public virtual void Drive()
    {
        Console.WriteLine("Транспортное средство находится в движении.");
    }
}

// Определение производного класса Car
public class Car : Vehicle
{
    public override void Drive()
    {
        Console.WriteLine("Машина едет.");
    }
}

// Определение производного класса Motorcycle
public class Motorcycle : Vehicle
{
    public override void Drive()
    {
        Console.WriteLine("Мотоцикл едет.");
    }
}

// Определение генерического класса VehicleCollection<T>
public class VehicleCollection<T> where T : Vehicle
{
    private List<T> _vehicles = new List<T>();

    public void Add(T vehicle)
    {
        _vehicles.Add(vehicle);
    }

    public void Remove(T vehicle)
    {
        _vehicles.Remove(vehicle);
    }

    public void DisplayVehicles()
    {
        foreach (var vehicle in _vehicles)
        {
            vehicle.Drive(); // Вызов метода Drive() для каждого транспортного средства
        }
    }
}


VehicleCollection<Car> cars = new VehicleCollection<Car>();
cars.Add(new Car());
cars.Add(new Motorcycle()); // Это возможно благодаря ограничению "where T : Vehicle"

cars.DisplayVehicles(); // Выводит: "Машина едет." и "Мотоцикл едет."


```

В этом примере `VehicleCollection<T>` является генерическим классом, который может хранить коллекцию объектов типа `T`, где `T` должен быть подклассом `Vehicle`. Это позволяет создавать коллекции автомобилей, мотоциклов и других видов транспортных средств, обеспечивая при этом строгую типизацию и безопасность типов.

### Заключение

Генерические классы в C# предоставляют мощный и гибкий способ создания компонентов, которые могут работать с различными типами данных, обеспечивая при этом строгую типизацию и безопасность типов. Это делает их идеальным выбором для многих задач, требующих повторного использования кода с высокой степенью абстракции.

<h4 style="color:DodgerBlue">Для проверки напишите пример кода на основе классов Vehicle, Car и Motorcycle ниже в блоке:</h4>

----

````
using System;
using System.Collections.Generic;

// === ОПРЕДЕЛЕНИЕ БАЗОВОГО И ПРОИЗВОДНЫХ КЛАССОВ ===

// Базовый класс Vehicle
public class Vehicle
{
    public virtual void Drive()
    {
        Console.WriteLine("Транспортное средство находится в движении.");
    }
}

// Производный класс Car
public class Car : Vehicle
{
    public override void Drive()
    {
        Console.WriteLine("Машина едет.");
    }
}

// Производный класс Motorcycle
public class Motorcycle : Vehicle
{
    public override void Drive()
    {
        Console.WriteLine("Мотоцикл едет.");
    }
}

// === ОПРЕДЕЛЕНИЕ ДЖЕНЕРИК-КЛАССА С ОГРАНИЧЕНИЕМ ТИПА ===

// Generic-класс VehicleCollection<T>
// where T : Vehicle — это ограничение, которое означает, что тип T должен быть 
// либо классом Vehicle, либо любым его производным классом.
public class VehicleCollection<T> where T : Vehicle
{
    // Приватный список для хранения объектов типа T
    private List<T> _vehicles = new List<T>();

    // Метод для добавления транспортного средства в коллекцию
    public void Add(T vehicle)
    {
        _vehicles.Add(vehicle);
    }

    // Метод для удаления транспортного средства из коллекции
    public void Remove(T vehicle)
    {
        _vehicles.Remove(vehicle);
    }

    // Метод для отображения всех транспортных средств в коллекции
    public void DisplayVehicles()
    {
        Console.WriteLine($"--- Коллекция из {_vehicles.Count} транспортных средств ---");
        foreach (var vehicle in _vehicles)
        {
            // Благодаря полиморфизму будет вызвана соответствующая версия метода Drive()
            vehicle.Drive();
        }
        Console.WriteLine("------------------------------------------");
    }
}

// === ИСПОЛЬЗОВАНИЕ ДЖЕНЕРИК-КЛАССА ===

class Program
{
    static void Main(string[] args)
    {
        // Создаем коллекцию, предназначенную только для автомобилей (Car)
        VehicleCollection<Car> carGarage = new VehicleCollection<Car>();
        carGarage.Add(new Car());
        carGarage.Add(new Car());
        
        // Благодаря ограничению 'where T : Vehicle', мы можем добавить и Motorcycle,
        // так как Motorcycle является наследником Vehicle.
        carGarage.Add(new Motorcycle()); 

        carGarage.DisplayVehicles();
        // Вывод:
        // Машина едет.
        // Машина едет.
        // Мотоцикл едет.


        // Создаем отдельную коллекцию только для мотоциклов
        VehicleCollection<Motorcycle> motorcycleClub = new VehicleCollection<Motorcycle>();
        motorcycleClub.Add(new Motorcycle());
        motorcycleClub.Add(new Motorcycle());

        motorcycleClub.DisplayVehicles();
        // Вывод:
        // Мотоцикл едет.
        // Мотоцикл едет.


        // Можно создать коллекцию и для базового класса Vehicle
        VehicleCollection<Vehicle> mixedTransport = new VehicleCollection<Vehicle>();
        mixedTransport.Add(new Vehicle());
        mixedTransport.Add(new Car());
        mixedTransport.Add(new Motorcycle());

        mixedTransport.DisplayVehicles();
        // Вывод:
        // Транспортное средство находится в движении.
        // Машина едет.
        // Мотоцикл едет.


        // Следующий код вызовет ошибку компиляции, так как string не наследуется от Vehicle:
        // VehicleCollection<string> strings = new VehicleCollection<string>(); // ОШИБКА!

        Console.ReadKey();
    }
}
````

````
using System;
using System.Collections.Generic;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== СОЗДАНИЕ КОЛЛЕКЦИИ ТОЛЬКО ДЛЯ СОБАК ===");
AnimalCollection<Dog> dogPark = new AnimalCollection<Dog>();
dogPark.Add(new Dog("Шарик", 3, "черный", "Овчарка"));
dogPark.Add(new Dog("Бобик", 5, "коричневый", "Лабрадор"));

dogPark.DisplayAnimals();
dogPark.PerformSpecialActions();

Console.WriteLine("\n=== СОЗДАНИЕ КОЛЛЕКЦИИ ТОЛЬКО ДЛЯ КОШЕК ===");
AnimalCollection<Cat> catHouse = new AnimalCollection<Cat>();
catHouse.Add(new Cat("Мурка", 5, "рыжая", true));
catHouse.Add(new Cat("Барсик", 2, "серый", false));

catHouse.DisplayAnimals();
catHouse.PerformSpecialActions();

Console.WriteLine("\n=== СОЗДАНИЕ КОЛЛЕКЦИИ ТОЛЬКО ДЛЯ ПТИЦ ===");
AnimalCollection<Bird> aviary = new AnimalCollection<Bird>();
aviary.Add(new Bird("Кеша", 2, "зеленый", 0.8));
aviary.Add(new Bird("Гоша", 4, "желтый", 1.2));

aviary.DisplayAnimals();
aviary.PerformSpecialActions();

Console.WriteLine("\n=== СОЗДАНИЕ СМЕШАННОЙ КОЛЛЕКЦИИ ЖИВОТНЫХ ===");
// Можно создать коллекцию и для базового класса Animal
AnimalCollection<Animal> zoo = new AnimalCollection<Animal>();
zoo.Add(new Animal("Неизвестное животное", 1, "серый"));
zoo.Add(new Dog("Рекс", 4, "черный", "Немецкая овчарка"));
zoo.Add(new Cat("Луна", 3, "белая", true));
zoo.Add(new Bird("Попугай", 1, "разноцветный", 0.5));

zoo.DisplayAnimals();
zoo.PerformSpecialActions();

// Следующий код вызовет ошибку компиляции, так как string не наследуется от Animal:
// AnimalCollection<string> strings = new AnimalCollection<string>(); // ОШИБКА!

// --- КЛАССЫ ---

// Базовый класс Animal
public class Animal
{
    public string Name { get; set; }
    public int Age { get; set; }
    public string Color { get; set; }

    public Animal(string name, int age, string color)
    {
        Name = name;
        Age = age;
        Color = color;
    }

    public virtual void MakeSound()
    {
        Console.WriteLine($"[Animal] {Name} издает какой-то звук.");
    }

    public void Sleep()
    {
        Console.WriteLine($"[Animal] {Name} спит.");
    }
}

// Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string color, string breed) : base(name, age, color)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"[Dog] {Name} ({Breed}) лает: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"[Dog] {Name} приносит палку.");
    }
}

// Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, string color, bool isIndoor) : base(name, age, color)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        string location = IsIndoor ? "дома" : "на улице";
        Console.WriteLine($"[Cat] {Name} мяукает: Мяу-мяу! (живет {location})");
    }

    public void Purr()
    {
        Console.WriteLine($"[Cat] {Name} мурлычет.");
    }
}

// Производный класс Bird
public class Bird : Animal
{
    public double Wingspan { get; set; }

    public Bird(string name, int age, string color, double wingspan) : base(name, age, color)
    {
        Wingspan = wingspan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"[Bird] {Name} щебечет: Чик-чирик! (размах крыльев: {Wingspan} м)");
    }

    public void Fly()
    {
        Console.WriteLine($"[Bird] {Name} летит высоко в небе.");
    }
}

// Generic-класс AnimalCollection<T> с ограничением типа
public class AnimalCollection<T> where T : Animal
{
    private List<T> _animals = new List<T>();

    public void Add(T animal)
    {
        _animals.Add(animal);
    }

    public void Remove(T animal)
    {
        _animals.Remove(animal);
    }

    public void DisplayAnimals()
    {
        Console.WriteLine($"--- Коллекция из {_animals.Count} животных ---");
        foreach (var animal in _animals)
        {
            animal.MakeSound();
            animal.Sleep();
            Console.WriteLine();
        }
        Console.WriteLine("------------------------------------------");
    }

    public void PerformSpecialActions()
    {
        Console.WriteLine($"--- Выполнение специальных действий для {_animals.Count} животных ---");
        foreach (var animal in _animals)
        {
            if (animal is Dog dog)
            {
                dog.Fetch();
            }
            else if (animal is Cat cat)
            {
                cat.Purr();
            }
            else if (animal is Bird bird)
            {
                bird.Fly();
            }
            else
            {
                Console.WriteLine($"[Animal] {animal.Name} не имеет специальных действий.");
            }
        }
        Console.WriteLine("------------------------------------------");
    }
}
````
