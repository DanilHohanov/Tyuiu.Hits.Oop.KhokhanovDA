<h1 style="color:DodgerBlue">Понятие перекрытия и перегрузка методов</h1>

### Перекрытие методов в C#

Перекрытие методов (или переопределение) — это процесс, при котором производный класс предоставляет новую реализацию метода, уже существующего в его базовом классе. Это позволяет изменить или расширить поведение метода в производном классе без изменения исходного метода в базовом классе. Перекрытие методов является ключевой особенностью полиморфизма в объектно-ориентированном программировании.

#### Пример с классами `Vehicle`, `Car` и `Motorcycle`

Давайте рассмотрим пример с классами `Vehicle`, `Car` и `Motorcycle`, где каждый из этих классов имеет метод `Drive()`, который принимает параметр `speed`.

```csharp
// Базовый класс Vehicle с виртуальным методом Drive()
public class Vehicle
{
    public virtual void Drive(int speed)
    {
        Console.WriteLine($"Транспорт движется {speed} km/h.");
    }
}

// Производный класс Car, переопределяющий метод Drive()
public class Car : Vehicle
{
    public override void Drive(int speed)
    {
        Console.WriteLine($"Автомобиль движется {speed} km/h.");
    }
}

// Производный класс Motorcycle, также переопределяющий метод Drive()
public class Motorcycle : Vehicle
{
    public override void Drive(int speed)
    {
        Console.WriteLine($"Мотоцикл движется {speed} km/h.");
    }
}
```

В этом примере, метод `Drive()` в классе `Car` и `Motorcycle` переопределяет метод `Drive()` из базового класса `Vehicle`. Это означает, что при вызове метода `Drive()` на объекте типа `Car` или `Motorcycle`, будет вызвана соответствующая реализация метода в этих классах, а не в классе `Vehicle`.

#### Отличие перекрытия от перегрузки методов

Перегрузка методов — это процесс определения нескольких методов с одним и тем же именем, но с разным количеством или типом параметров. Например:

```csharp
public class ExampleClass
{
    public void Print(int number)
    {
        Console.WriteLine(number);
    }

    public void Print(double number)
    {
        Console.WriteLine(number.ToString());
    }
}
```

В этом случае, у нас есть два метода с именем `Print`, но с разным типом параметра. Это отличается от перекрытия методов, где метод в производном классе полностью заменяет метод в базовом классе.

#### Вызов методов родителя

Иногда в производном классе может потребоваться вызвать метод родителя. Для этого используется ключевое слово `base`. Например, если мы хотим добавить дополнительную логику в метод `Drive()` класса `Car`, сохраняя при этом оригинальное поведение метода `Drive()` из класса `Vehicle`.

```csharp
public class Car : Vehicle
{
    public override void Drive(int speed)
    {
        base.Drive(speed); // Вызов метода Drive() из класса Vehicle
        Console.WriteLine("Добавляем специфическое поведение автомобиля");
    }
}
```

В этом примере, после вызова `base.Drive(speed)`, выполняется дополнительная логика, специфичная для класса `Car`.


### Перегрузка методов в C#

Перегрузка методов — это способность класса определить несколько методов с одним и тем же именем, но с разным количеством или типом параметров. Это позволяет разработчикам предоставлять различные варианты выполнения одного и того же действия в зависимости от входных данных. Перегрузка методов является одной из форм выражения полиморфизма в объектно-ориентированном программировании.

#### Пример с классами `Vehicle`, `Car` и `Motorcycle`

Давайте рассмотрим пример, где класс `Vehicle` содержит перегруженные методы `Drive()`, которые принимают разные типы параметров.

```csharp
public class Vehicle
{
    // Метод Drive(), принимающий int
    public void Drive(int speed)
    {
        Console.WriteLine($"Транспорт движется {speed} km/h.");
    }

    // Метод Drive(), принимающий string
    public void Drive(string mode)
    {
        Console.WriteLine($"Автомобиль движется в {mode} режиме.");
    }

    // Метод Drive(), принимающий bool
    public void Drive(bool isManual)
    {
        if (isManual)
        {
            Console.WriteLine("Автомобиль переключает передачи вручную.");
        }
        else
        {
            Console.WriteLine("Автомобиль оснащен автоматической коробкой передач.");
        }
    }
}
```

В этом примере класс `Vehicle` определяет три метода с именем `Drive()`, каждый из которых принимает разные типы параметров: `int`, `string` и `bool`. Это позволяет вызывать разные версии метода `Drive()` в зависимости от контекста, например:

```csharp
Vehicle myVehicle = new Vehicle();

myVehicle.Drive(60); // Вызов метода Drive() с int параметром
myVehicle.Drive("Eco"); // Вызов метода Drive() с string параметром
myVehicle.Drive(true); // Вызов метода Drive() с bool параметром
```

#### Отличие перегрузки от переопределения

Отличие перегрузки методов от переопределения заключается в том, что перегрузка методов включает определение нескольких методов с одним именем, но с разным набором параметров, тогда как переопределение методов (или перекрытие) означает предоставление новой реализации метода в производном классе, который уже существует в базовом классе.

#### Вызов методов

При вызове перегруженного метода компилятор выбирает подходящую версию метода на основе типов и количества аргументов, переданных в вызове. Если типы и количество аргументов совпадают с несколькими версиями метода, компилятор выберет ту версию, которая была объявлена первой.

### Заключение

Перегрузка методов в C# позволяет классам предоставлять различные варианты выполнения одного и того же действия, адаптируясь к различным условиям использования. Это мощный инструмент для создания гибких и удобных API, позволяющих клиентам класса использовать его функциональность наиболее эффективно.

Перекрытие методов в C# позволяет производным классам изменять или расширять поведение методов базовых классов, что является ключевой особенностью полиморфизма. Это отличается от перегрузки методов, которая предполагает определение нескольких методов с одинаковым именем, но с разным набором параметров.



<h4 style="color:DodgerBlue">Для проверки напишите пример кода на основе классов Vehicle, Car и Motorcycle ниже в блоке:</h4>

----

````
using System;

// Определение классов с перекрытием и перегрузкой методов

// Базовый класс Vehicle
public class Vehicle
{
    // --- Перегрузка методов (Overloading) ---
    // Это несколько методов с одним именем, но разными параметрами.
    // Они существуют только в этом классе.

    public void Drive(int speed)
    {
        Console.WriteLine($"[Vehicle] Транспорт движется {speed} км/ч.");
    }

    public void Drive(string mode)
    {
        Console.WriteLine($"[Vehicle] Транспорт движется в режиме '{mode}'.");
    }

    public void Drive(bool isManual)
    {
        if (isManual)
        {
            Console.WriteLine("[Vehicle] Управление осуществляется вручную.");
        }
        else
        {
            Console.WriteLine("[Vehicle] Используется автоматическое управление.");
        }
    }

    // --- Виртуальный метод для перекрытия (Overriding) ---
    // Этот метод предназначен для переопределения в производных классах.
    public virtual void MakeSound()
    {
        Console.WriteLine("[Vehicle] Издает общий звук транспорта.");
    }
}

// Производный класс Car
public class Car : Vehicle
{
    // --- Перекрытие метода (Overriding) ---
    // Мы предоставляем новую реализацию для виртуального метода базового класса.
    public override void MakeSound()
    {
        Console.WriteLine("[Car] Сигналит: Би-би!");
    }

    // --- Вызов метода родителя через base ---
    // Мы можем расширить поведение метода базового класса, а не полностью его заменять.
    public override void Drive(int speed)
    {
        // Сначала вызываем оригинальную реализацию из класса Vehicle
        base.Drive(speed);
        // Затем добавляем свою собственную логику
        Console.WriteLine("[Car] Автомобиль плавно разгоняется.");
    }

    // Car также наследует перегруженные методы Drive(string) и Drive(bool) от Vehicle.
    // Мы можем их использовать как есть или тоже перекрыть.
}

// Производный класс Motorcycle
public class Motorcycle : Vehicle
{
    // Полностью заменяем поведение метода MakeSound()
    public override void MakeSound()
    {
        Console.WriteLine("[Motorcycle] Рычит: Врум-врум!");
    }

    // Переопределяем метод Drive(int), но НЕ вызываем base.Drive().
    // Это демонстрирует полную замену поведения.
    public override void Drive(int speed)
    {
        Console.WriteLine($"[Motorcycle] Мотоцикл мчится со скоростью {speed} км/ч.");
    }
}

class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== Демонстрация перегрузки методов (Overloading) ===");
        Vehicle myVehicle = new Vehicle();
        
        // Вызываем разные версии метода Drive() в зависимости от типа параметра
        myVehicle.Drive(60);      // Вызывает Drive(int)
        myVehicle.Drive("Спорт"); // Вызывает Drive(string)
        myVehicle.Drive(true);    // Вызывает Drive(bool)
        Console.WriteLine();

        Console.WriteLine("=== Демонстрация перекрытия методов (Overriding) и полиморфизма ===");
        // Создаем список базового типа, но заполняем его производными объектами
        Vehicle[] vehicles = {
            new Car(),
            new Motorcycle(),
            new Vehicle()
        };

        foreach (var vehicle in vehicles)
        {
            // При вызове MakeSound() будет выполнена версия, соответствующая
            // реальному типу объекта (Car, Motorcycle или Vehicle).
            vehicle.MakeSound();
        }
        Console.WriteLine();

        Console.WriteLine("=== Демонстрация вызова метода родителя через 'base' ===");
        Car myCar = new Car();
        // При вызове Drive(int) у объекта Car, сначала выполнится код из Vehicle,
        // а затем дополнительный код из Car.
        myCar.Drive(100);
        Console.WriteLine();

        Console.WriteLine("=== Демонстрация полной замены метода в производном классе ===");
        Motorcycle myMotorcycle = new Motorcycle();
        // Здесь вызывается только версия метода из класса Motorcycle.
        myMotorcycle.Drive(120);

        Console.ReadKey();
    }
}
````

Ниже в блоке по примеру создайте базовый класс Animal и производные классы (3-4 например Dog, Cat, Bird и так далее) реализуйте структуру и объявление класса, включая свойства, геттеры и сеттеры, а также используйте перекрытия и перегрузки методов.

````
using System;
using System.Collections.Generic;

// === БАЗОВЫЙ КЛАСС ANIMAL ===
public class Animal
{
    // Свойства с геттерами и сеттерами
    public string Name { get; set; }
    public int Age { get; set; }
    public string Color { get; set; }

    // Конструктор для удобства инициализации
    public Animal(string name, int age, string color)
    {
        Name = name;
        Age = age;
        Color = color;
    }

    // --- ПЕРЕГРУЗКА МЕТОДОВ (Overloading) ---
    // Несколько методов с одним именем, но разными параметрами

    // Версия 1: принимает скорость
    public void Move(int speed)
    {
        Console.WriteLine($"[Animal] {Name} движется со скоростью {speed} км/ч.");
    }

    // Версия 2: принимает направление
    public void Move(string direction)
    {
        Console.WriteLine($"[Animal] {Name} движется в направлении: {direction}.");
    }

    // Версия 3: принимает расстояние и время
    public void Move(double distance, double timeHours)
    {
        double avgSpeed = distance / timeHours;
        Console.WriteLine($"[Animal] {Name} прошел {distance} км за {timeHours} ч. Средняя скорость: {avgSpeed:F1} км/ч.");
    }

    // --- ВИРТУАЛЬНЫЙ МЕТОД ДЛЯ ПЕРЕКРЫТИЯ (Overriding) ---
    public virtual void MakeSound()
    {
        Console.WriteLine($"[Animal] {Name} издает какой-то звук.");
    }

    // Обычный метод, доступный всем животным
    public void Sleep()
    {
        Console.WriteLine($"[Animal] {Name} спит.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС DOG ===
public class Dog : Animal
{
    // Дополнительное свойство
    public string Breed { get; set; }

    public Dog(string name, int age, string color, string breed) : base(name, age, color)
    {
        Breed = breed;
    }

    // --- ПЕРЕКРЫТИЕ МЕТОДА (Overriding) ---
    public override void MakeSound()
    {
        Console.WriteLine($"[Dog] {Name} ({Breed}) лает: Гав-гав!");
    }

    // Переопределяем одну из перегруженных версий метода Move
    public override void Move(int speed)
    {
        // Вызываем реализацию базового класса через base
        base.Move(speed);
        // Добавляем специфичное поведение
        Console.WriteLine($"[Dog] {Name} радостно виляет хвостом при беге!");
    }

    // Специфичный для собаки метод
    public void Fetch()
    {
        Console.WriteLine($"[Dog] {Name} приносит палку.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС CAT ===
public class Cat : Animal
{
    // Дополнительное свойство
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, string color, bool isIndoor) : base(name, age, color)
    {
        IsIndoor = isIndoor;
    }

    // Перекрытие метода MakeSound
    public override void MakeSound()
    {
        string location = IsIndoor ? "дома" : "на улице";
        Console.WriteLine($"[Cat] {Name} мяукает: Мяу-мяу! (живет {location})");
    }

    // Полностью заменяем поведение метода Move(int) без вызова base
    public override void Move(int speed)
    {
        Console.WriteLine($"[Cat] {Name} крадется со скоростью {speed} км/ч... очень тихо.");
    }

    // Специфичный для кошки метод
    public void Purr()
    {
        Console.WriteLine($"[Cat] {Name} мурлычет.");
    }
}

// === ПРОИЗВОДНЫЙ КЛАСС BIRD ===
public class Bird : Animal
{
    // Дополнительное свойство
    public double Wingspan { get; set; } // размах крыльев в метрах

    public Bird(string name, int age, string color, double wingspan) : base(name, age, color)
    {
        Wingspan = wingspan;
    }

    // Перекрытие метода MakeSound
    public override void MakeSound()
    {
        Console.WriteLine($"[Bird] {Name} щебечет: Чик-чирик! (размах крыльев: {Wingspan} м)");
    }

    // Переопределяем другую версию метода Move — ту, что принимает направление
    public override void Move(string direction)
    {
        Console.WriteLine($"[Bird] {Name} летит в направлении: {direction}, взмахивая крыльями.");
    }

    // Специфичный для птицы метод
    public void Fly()
    {
        Console.WriteLine($"[Bird] {Name} парит высоко в небе.");
    }
}

// === ИСПОЛЬЗОВАНИЕ ПОЛИМОРФИЗМА ===
class Program
{
    static void Main(string[] args)
    {
        Console.WriteLine("=== ДЕМОНСТРАЦИЯ ПЕРЕГРУЗКИ МЕТОДОВ (OVERLOADING) ===");
        Animal myAnimal = new Animal("Неизвестное животное", 1, "серый");
        
        // Вызываем разные версии метода Move() в зависимости от типа параметра
        myAnimal.Move(50);                    // Вызывает Move(int)
        myAnimal.Move("на север");            // Вызывает Move(string)
        myAnimal.Move(10.5, 2.0);             // Вызывает Move(double, double)
        Console.WriteLine();

        Console.WriteLine("=== ДЕМОНСТРАЦИЯ ПЕРЕКРЫТИЯ МЕТОДОВ (OVERRIDING) И ПОЛИМОРФИЗМА ===");
        // Создаем список базового типа, но заполняем его производными объектами
        List<Animal> animals = new List<Animal>
        {
            new Dog("Шарик", 3, "черный", "Овчарка"),
            new Cat("Мурка", 5, "рыжая", true),
            new Bird("Кеша", 2, "зеленый", 0.8),
            new Animal("Неизвестное животное", 1, "серый")
        };

        foreach (var animal in animals)
        {
            // При вызове MakeSound() будет выполнена версия, соответствующая
            // реальному типу объекта (полиморфизм).
            animal.MakeSound();
            
            // Вызываем обычный метод базового класса
            animal.Sleep();
            
            Console.WriteLine(); // Пустая строка для разделения
        }

        Console.WriteLine("=== ДЕМОНСТРАЦИЯ СПЕЦИФИЧНЫХ МЕТОДОВ И ЧАСТИЧНОГО ПЕРЕКРЫТИЯ ===");
        // Чтобы вызвать специфичные методы или проверить частичное перекрытие,
        // нужно привести тип к конкретному классу
        foreach (var animal in animals)
        {
            if (animal is Dog dog)
            {
                dog.Fetch();
                // Проверим, как работает частично перекрытый метод Move(int)
                dog.Move(40); // Здесь вызовется overridden метод, который сначала вызовет base.Move(40)
            }
            else if (animal is Cat cat)
            {
                cat.Purr();
                cat.Move(30); // Здесь вызовется полностью замененный метод
            }
            else if (animal is Bird bird)
            {
                bird.Fly();
                bird.Move("в южном направлении"); // Здесь вызовется overridden метод Move(string)
            }
        }

        Console.ReadKey();
    }
}
````
