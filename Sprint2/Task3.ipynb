<h2 style="color:DodgerBlue">Описание проекта:</h2>

----

Создать базовый класс Invoice в C#, который будет представлять информацию о фактурах за поставленные товары или оказанные услуги. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для
демонстрации полиморфизма.
Требования к базовому классу Invoice:

• Атрибуты: Номер фактуры (InvoiceNumber), Дата выдачи (IssueDate), Общая
сумма (TotalAmount).

• Методы:

o CalculateTotal(): метод для расчета общей суммы по фактуре.

o AddLine(LineItem lineItem): метод для добавления позиции в фактуру.

o RemoveLine(LineItem lineItem): метод для удаления позиции из фактуры.

Требования к производным классам:

1. ТоварнаяФактура (GoodsInvoice): Должна содержать дополнительные атрибуты, такие как Дата поставки (SupplyDate). Метод AddLine() должен быть переопределен для добавления информации о дате поставки товара при добавлении позиции.

2. УслуговаяФактура (ServiceInvoice): Должна содержать дополнительные атрибуты, такие как Дата оказания услуги (ServiceDate). Метод RemoveLine() должен быть переопределен для добавления информации о причине аннулирования услуги при удалении позиции.

3. КомбинированнаяФактура (CombinedInvoice) (если требуется третий класс): Должна содержать дополнительные атрибуты, такие как Наличие возврата (ReturnAllowed). Метод CalculateTotal() должен быть переопределен для учета возможного возврата товара или услуги при расчете общей суммы.

#### Дополнительное задание
Добавьте к сущестующим классам (базовыму и производным 3-4 атрибута и метода) и реализуйте простое, сложное и множественное наследование

```
using System;
using System.Collections.Generic;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== ДЕМОНСТРАЦИЯ НАСЛЕДОВАНИЯ И ПОЛИМОРФИЗМА ===\n");

// --- Простое наследование: GoodsInvoice ---
var goodsInv = new GoodsInvoice("INV-001", DateTime.Now, DateTime.Now.AddDays(3), "ООО 'Поставщик'");
goodsInv.WarehouseLocation = "Склад №5";
goodsInv.AddLine(new LineItem("Ноутбук", 1, 75000));
goodsInv.AddLine(new LineItem("Мышь", 2, 1500));
goodsInv.GenerateDeliveryNote();
goodsInv.PrintSummary();
goodsInv.MarkAsPaid();

Console.WriteLine("\n" + new string('-', 60) + "\n");

// --- Сложное наследование: PremiumServiceInvoice ---
var premiumServInv = new PremiumServiceInvoice("SRV-002", DateTime.Now, DateTime.Now, "Ремонт компьютера", true);
premiumServInv.TechnicianName = "Иван Петров";
premiumServInv.ServiceDurationHours = 4;
premiumServInv.DiscountPercentage = 10;
premiumServInv.WarrantyPeriod = "1 год";
premiumServInv.AddLine(new LineItem("Диагностика", 1, 2000));
premiumServInv.AddLine(new LineItem("Замена HDD", 1, 5000));
premiumServInv.ScheduleFollowUp();
premiumServInv.ActivateWarranty();
premiumServInv.PrintSummary();
premiumServInv.CalculateTotal(); // Покажет применение скидки

Console.WriteLine("\n" + new string('-', 60) + "\n");

// --- Множественное наследование: CombinedInvoice ---
var combinedInv = new CombinedInvoice("COMB-003", DateTime.Now, true);
combinedInv.ReturnDeadline = DateTime.Now.AddDays(14);
combinedInv.TaxRate = 0.15m; // 15%
combinedInv.AddLine(new LineItem("Консультация", 2, 3000));
combinedInv.AddLine(new LineItem("Лицензия ПО", 1, 15000));
combinedInv.CollectFeedback(); // Автоматический отзыв
combinedInv.PrintSummary();
combinedInv.CalculateTotal();
combinedInv.CalculateTax(); // Применяем налог

// Создаем существующую позицию для возврата
var consultationItem = new LineItem("Консультация", 2, 3000);
combinedInv.AddLine(consultationItem); // Добавляем, чтобы можно было вернуть
combinedInv.ProcessReturn(consultationItem); // Возвращаем позицию

Console.WriteLine("\n" + new string('-', 60) + "\n");

// --- Полиморфизм: работа с коллекцией базовых типов ---
List<Invoice> invoices = new List<Invoice>
{
    goodsInv,
    premiumServInv,
    combinedInv
};

Console.WriteLine("=== ПОЛИМОРФИЗМ: Вызов методов через базовый тип ===");
foreach (var inv in invoices)
{
    inv.PrintSummary();
    Console.WriteLine(new string('-', 40));
}

// --- КЛАССЫ ---

// ========================
// БАЗОВЫЙ КЛАСС: Invoice
// ========================
public abstract class Invoice
{
    public string InvoiceNumber { get; set; }
    public DateTime IssueDate { get; set; }
    public decimal TotalAmount { get; protected set; }
    public List<LineItem> Lines { get; set; } = new List<LineItem>();
    public string Currency { get; set; } = "RUB";
    public bool IsPaid { get; set; } = false;

    protected Invoice(string invoiceNumber, DateTime issueDate)
    {
        InvoiceNumber = invoiceNumber;
        IssueDate = issueDate;
    }

    public virtual void CalculateTotal()
    {
        TotalAmount = 0;
        foreach (var line in Lines)
        {
            TotalAmount += line.Quantity * line.Price;
        }
        Console.WriteLine($"Общая сумма по фактуре {InvoiceNumber}: {TotalAmount} {Currency}");
    }

    public virtual void AddLine(LineItem lineItem)
    {
        Lines.Add(lineItem);
        Console.WriteLine($"Добавлена позиция: {lineItem.Description} x {lineItem.Quantity}");
    }

    public virtual void RemoveLine(LineItem lineItem)
    {
        if (Lines.Remove(lineItem))
        {
            Console.WriteLine($"Удалена позиция: {lineItem.Description}");
        }
        else
        {
            Console.WriteLine("Позиция не найдена.");
        }
    }

    public void MarkAsPaid()
    {
        IsPaid = true;
        Console.WriteLine($"Фактура {InvoiceNumber} помечена как оплаченная.");
    }

    public virtual void PrintSummary()
    {
        Console.WriteLine($"\n=== Фактура № {InvoiceNumber} ===");
        Console.WriteLine($"Дата выдачи: {IssueDate:d}");
        Console.WriteLine($"Валюта: {Currency}");
        Console.WriteLine($"Статус оплаты: {(IsPaid ? "Оплачена" : "Не оплачена")}");
        Console.WriteLine($"Количество позиций: {Lines.Count}");
        CalculateTotal();
    }
}

// ========================
// КЛАСС: GoodsInvoice
// ========================
public class GoodsInvoice : Invoice
{
    public DateTime SupplyDate { get; set; }
    public string SupplierName { get; set; }
    public string WarehouseLocation { get; set; }

    public GoodsInvoice(string invoiceNumber, DateTime issueDate, DateTime supplyDate, string supplierName) 
        : base(invoiceNumber, issueDate)
    {
        SupplyDate = supplyDate;
        SupplierName = supplierName;
    }

    public override void AddLine(LineItem lineItem)
    {
        base.AddLine(lineItem);
        Console.WriteLine($"[GoodsInvoice] Товар поставлен: {SupplyDate:d}, Поставщик: {SupplierName}");
    }

    public void GenerateDeliveryNote()
    {
        Console.WriteLine($"Генерация накладной для поставки от {SupplyDate:d} от поставщика {SupplierName}");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"Тип: Товарная фактура");
        Console.WriteLine($"Дата поставки: {SupplyDate:d}");
        Console.WriteLine($"Поставщик: {SupplierName}");
        Console.WriteLine($"Склад: {WarehouseLocation ?? "Не указан"}");
    }
}

// ========================
// КЛАСС: ServiceInvoice
// ========================
public class ServiceInvoice : Invoice
{
    public DateTime ServiceDate { get; set; }
    public string ServiceDescription { get; set; }
    public string TechnicianName { get; set; }
    public int ServiceDurationHours { get; set; }

    public ServiceInvoice(string invoiceNumber, DateTime issueDate, DateTime serviceDate, string description) 
        : base(invoiceNumber, issueDate)
    {
        ServiceDate = serviceDate;
        ServiceDescription = description;
    }

    public override void RemoveLine(LineItem lineItem)
    {
        if (Lines.Remove(lineItem))
        {
            Console.WriteLine($"[ServiceInvoice] Удалена услуга: {lineItem.Description}. Причина: Отмена заказа клиентом.");
        }
        else
        {
            Console.WriteLine("Услуга не найдена.");
        }
    }

    public void ScheduleFollowUp()
    {
        Console.WriteLine($"Запланировано повторное обслуживание для услуги '{ServiceDescription}' через 30 дней.");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"Тип: Услуговая фактура");
        Console.WriteLine($"Дата оказания услуги: {ServiceDate:d}");
        Console.WriteLine($"Описание услуги: {ServiceDescription}");
        Console.WriteLine($"Техник: {TechnicianName ?? "Не назначен"}");
        Console.WriteLine($"Продолжительность: {ServiceDurationHours} ч.");
    }
}

// ========================
// КЛАСС: PremiumServiceInvoice
// ========================
public class PremiumServiceInvoice : ServiceInvoice
{
    public bool IncludesWarranty { get; set; }
    public string WarrantyPeriod { get; set; }
    public decimal DiscountPercentage { get; set; }

    public PremiumServiceInvoice(string invoiceNumber, DateTime issueDate, DateTime serviceDate, string description, bool includesWarranty) 
        : base(invoiceNumber, issueDate, serviceDate, description)
    {
        IncludesWarranty = includesWarranty;
    }

    public override void CalculateTotal()
    {
        base.CalculateTotal();
        if (DiscountPercentage > 0)
        {
            decimal discount = TotalAmount * (DiscountPercentage / 100);
            TotalAmount -= discount;
            Console.WriteLine($"[PremiumServiceInvoice] Применена скидка {DiscountPercentage}%: -{discount} {Currency}");
        }
        Console.WriteLine($"Итоговая сумма после скидки: {TotalAmount} {Currency}");
    }

    public void ActivateWarranty()
    {
        if (IncludesWarranty)
        {
            Console.WriteLine($"Активирована гарантия на {WarrantyPeriod} для услуги '{ServiceDescription}'.");
        }
        else
        {
            Console.WriteLine("Гарантия не предусмотрена для этой услуги.");
        }
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"Тип: Премиум услуговая фактура");
        Console.WriteLine($"Гарантия: {(IncludesWarranty ? $"Да ({WarrantyPeriod})" : "Нет")}");
        Console.WriteLine($"Скидка: {DiscountPercentage}%");
    }
}

// ========================
// ИНТЕРФЕЙСЫ
// ========================
public interface IReturnable
{
    bool ReturnAllowed { get; set; }
    void ProcessReturn(LineItem item);
}

public interface ITaxCalculable
{
    decimal TaxRate { get; set; }
    void CalculateTax();
}

// ========================
// КЛАСС: CombinedInvoice
// ========================
public class CombinedInvoice : Invoice, IReturnable, ITaxCalculable
{
    public bool ReturnAllowed { get; set; }
    public decimal TaxRate { get; set; } = 0.2m;
    public DateTime ReturnDeadline { get; set; }
    public string CustomerFeedback { get; set; }

    public CombinedInvoice(string invoiceNumber, DateTime issueDate, bool returnAllowed) 
        : base(invoiceNumber, issueDate)
    {
        ReturnAllowed = returnAllowed;
    }

    public void ProcessReturn(LineItem item)
    {
        if (ReturnAllowed && Lines.Contains(item))
        {
            Lines.Remove(item);
            Console.WriteLine($"[CombinedInvoice] Обработан возврат товара/услуги: {item.Description}");
            CalculateTotal();
        }
        else
        {
            Console.WriteLine("Возврат невозможен или позиция не найдена.");
        }
    }

    public void CalculateTax()
    {
        decimal taxAmount = TotalAmount * TaxRate;
        Console.WriteLine($"[CombinedInvoice] Налог ({TaxRate * 100}%): {taxAmount} {Currency}");
        TotalAmount += taxAmount;
        Console.WriteLine($"Итоговая сумма с налогом: {TotalAmount} {Currency}");
    }

    public override void CalculateTotal()
    {
        base.CalculateTotal();
        if (ReturnAllowed)
        {
            Console.WriteLine("[CombinedInvoice] Учтена возможность возврата при расчете суммы.");
        }
    }

    public void CollectFeedback()
    {
        CustomerFeedback = "Отличное обслуживание, рекомендую!";
        Console.WriteLine($"[CombinedInvoice] Отзыв клиента: \"{CustomerFeedback}\"");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"Тип: Комбинированная фактура");
        Console.WriteLine($"Возврат разрешен: {ReturnAllowed}");
        Console.WriteLine($"Ставка налога: {TaxRate * 100}%");
        Console.WriteLine($"Крайний срок возврата: {ReturnDeadline:d}");
        Console.WriteLine($"Отзыв клиента: {CustomerFeedback ?? "Не собран"}");
    }
}

// ========================
// КЛАСС: LineItem
// ========================
public class LineItem
{
    public string Description { get; set; }
    public int Quantity { get; set; }
    public decimal Price { get; set; }

    public LineItem(string description, int quantity, decimal price)
    {
        Description = description;
        Quantity = quantity;
        Price = price;
    }

    // Для корректной работы Contains и Remove
    public override bool Equals(object obj)
    {
        return obj is LineItem other && 
               Description == other.Description && 
               Quantity == other.Quantity && 
               Price == other.Price;
    }

    public override int GetHashCode()
    {
        return HashCode.Combine(Description, Quantity, Price);
    }
}
```
