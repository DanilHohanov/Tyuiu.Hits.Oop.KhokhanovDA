### Интерфейсы в C#

**Интерфейс** — это контракт, который определяет набор методов и свойств, которые класс должен реализовать. Интерфейсы позволяют создать гибкую и расширяемую архитектуру, поскольку они отделяют определение поведения от его реализации.

Давайте подробно разберем понятия интерфейсов и управления зависимостями в C# на примере классов `Vehicle`, `Car` и `Motorcycle`.

#### Пример интерфейса

Предположим, мы имеем интерфейс `IVehicle`, который определяет поведение транспортных средств. Ниже приведен пример его определения:

```csharp
public interface IVehicle
{
    void Start();            // Метод для запуска транспортного средства
    void Stop();             // Метод для остановки транспортного средства
    int GetNumberOfWheels(); // Метод для получения количества колес
}
```

Интерфейс `IVehicle` требует, чтобы любой класс, реализующий его, обеспечил реализацию этих методов.

### Реализация интерфейса

Теперь создадим классы `Car` и `Motorcycle`, которые будут реализовывать этот интерфейс:

```csharp
public class Car : IVehicle
{
    public void Start()
    {
        Console.WriteLine("Машина запущена.");
    }

    public void Stop()
    {
        Console.WriteLine("Машина остановлена.");
    }

    public int GetNumberOfWheels()
    {
        return 4; // У машины 4 колеса
    }
}

public class Motorcycle : IVehicle
{
    public void Start()
    {
        Console.WriteLine("Мотоцикл запущен.");
    }

    public void Stop()
    {
        Console.WriteLine("Мотоцикл остановлен.");
    }

    public int GetNumberOfWheels()
    {
        return 2; // У мотоцикла 2 колеса
    }
}
```

Каждый класс (`Car` и `Motorcycle`) реализует все методы, определенные в интерфейсе `IVehicle`, что позволяет создать единый интерфейс для работы с разными типами транспортных средств.

### Управление зависимостями (Dependency Injection)

**Управление зависимостями** — это паттерн проектирования, который помогает управлять созданием зависимостей в классе. Этот подход позволяет внедрять (или "вводить") зависимости через конструктор, методы или свойства, вместо того чтобы создавать их внутри класса. Это делает код более чистым и облегчает его тестирование.

#### Пример управления зависимостями

Создадим класс `VehicleService`, который будет зависеть от интерфейса `IVehicle`:

```csharp
public class VehicleService
{
    private readonly IVehicle _vehicle;

    // Внедрение зависимости через конструктор
    public VehicleService(IVehicle vehicle)
    {
        _vehicle = vehicle; 
    }

    public void UseVehicle()
    {
        _vehicle.Start(); // Запускаем транспортное средство
        Console.WriteLine($"У этого транспортного средства {_vehicle.GetNumberOfWheels()} колеса.");
        _vehicle.Stop(); // Останавливаем транспортное средство
    }
}
```

В этом примере класс `VehicleService` принимает в своем конструкторе объект, реализующий интерфейс `IVehicle`. Это позволяет `VehicleService` работать с любым объектом, который реализует `IVehicle`, обеспечивая гибкость и уменьшая связывание между классами.

### Использование классов

Теперь давайте создадим объекты `Car` и `Motorcycle` и используем наш сервис:

```csharp
IVehicle myCar = new Car(); // Создаем объект машины
VehicleService carService = new VehicleService(myCar); // Передаем машину в сервис
carService.UseVehicle(); // Используем сервис с машиной

IVehicle myMotorcycle = new Motorcycle(); // Создаем объект мотоцикла
VehicleService motorcycleService = new VehicleService(myMotorcycle); // Передаем мотоцикл в сервис
motorcycleService.UseVehicle(); // Используем сервис с мотоциклом

```

### Результат выполнения

При запуске данной программы будет следующий вывод:

```
Машина запущена.
У этого транспортного средства 4 колеса.
Машина остановлена.
Мотоцикл запущен.
У этого транспортного средства 2 колеса.
Мотоцикл остановлен.
```

### Заключение

Использование интерфейсов и управления зависимостями позволяет создать более организованный и тестируемый код. Интерфейсы определяют стандартные поведения, а управление зависимостями снижает связность между компонентами. Это делает систему более гибкой и упрощает процесс её поддержки и расширения в будущем.

````
using System;

// 1. Определение интерфейса IVehicle
public interface IVehicle
{
    void Start();
    void Stop();
    int GetNumberOfWheels();
}

// 2. Абстрактный базовый класс Vehicle, реализующий IVehicle
public abstract class Vehicle : IVehicle
{
    public string Make { get; set; }
    public string Model { get; set; }

    public Vehicle(string make, string model)
    {
        Make = make;
        Model = model;
    }

    // Реализация методов интерфейса (могут быть переопределены в наследниках)
    public virtual void Start()
    {
        Console.WriteLine($"{Make} {Model} запущен.");
    }

    public virtual void Stop()
    {
        Console.WriteLine($"{Make} {Model} остановлен.");
    }

    // Абстрактный метод - должен быть реализован в каждом конкретном классе
    public abstract int GetNumberOfWheels();
}

// 3. Производный класс Car
public class Car : Vehicle
{
    public int NumberOfDoors { get; set; }

    public Car(string make, string model, int numberOfDoors) : base(make, model)
    {
        NumberOfDoors = numberOfDoors;
    }

    // Переопределение метода Start() для автомобиля
    public override void Start()
    {
        Console.WriteLine($"Автомобиль {Make} {Model} запущен.");
    }

    // Реализация абстрактного метода
    public override int GetNumberOfWheels()
    {
        return 4; // У машины 4 колеса
    }

    // Уникальный метод для автомобиля
    public void Honk()
    {
        Console.WriteLine("Бип-бип!");
    }
}

// 4. Производный класс Motorcycle
public class Motorcycle : Vehicle
{
    public bool HasSidecar { get; set; }

    public Motorcycle(string make, string model, bool hasSidecar) : base(make, model)
    {
        HasSidecar = hasSidecar;
    }

    // Переопределение метода Start() для мотоцикла
    public override void Start()
    {
        Console.WriteLine($"Мотоцикл {Make} {Model} запущен.");
    }

    // Реализация абстрактного метода
    public override int GetNumberOfWheels()
    {
        return HasSidecar ? 3 : 2; // У мотоцикла 2 или 3 колеса (с коляской)
    }

    // Уникальный метод для мотоцикла
    public void RevEngine()
    {
        Console.WriteLine("Рррррр!");
    }
}

// 5. Класс сервиса, использующий внедрение зависимостей
public class VehicleService
{
    private readonly IVehicle _vehicle;

    // Внедрение зависимости через конструктор
    public VehicleService(IVehicle vehicle)
    {
        _vehicle = vehicle;
    }

    public void UseVehicle()
    {
        _vehicle.Start(); // Запускаем транспортное средство
        Console.WriteLine($"У этого транспортного средства {_vehicle.GetNumberOfWheels()} колеса.");
        
        // Можно проверить тип и вызвать уникальные методы
        if (_vehicle is Car car)
        {
            car.Honk();
        }
        else if (_vehicle is Motorcycle motorcycle)
        {
            motorcycle.RevEngine();
        }
        
        _vehicle.Stop(); // Останавливаем транспортное средство
    }

    public void DisplayInfo()
    {
        if (_vehicle is Vehicle v)
        {
            Console.WriteLine($"Марка: {v.Make}, Модель: {v.Model}");
        }
    }
}

// 6. Пример использования всех классов и сервисов
class Program
{
    static void Main()
    {
        Console.WriteLine("=== Демонстрация внедрения зависимостей ===");

        // Создание объектов
        IVehicle myCar = new Car("Toyota", "Camry", 4);
        IVehicle myMotorcycle = new Motorcycle("Harley-Davidson", "Street", false);

        // Создание сервисов с внедрением зависимостей
        VehicleService carService = new VehicleService(myCar);
        VehicleService motorcycleService = new VehicleService(myMotorcycle);

        // Использование сервисов
        Console.WriteLine("--- Работа с автомобилем ---");
        carService.DisplayInfo();
        carService.UseVehicle();

        Console.WriteLine();
        Console.WriteLine("--- Работа с мотоциклом ---");
        motorcycleService.DisplayInfo();
        motorcycleService.UseVehicle();

        Console.WriteLine();
        Console.WriteLine("=== Полиморфизм: коллекция транспортных средств ===");

        // Коллекция объектов типа IVehicle
        IVehicle[] vehicles = new IVehicle[]
        {
            new Car("BMW", "X5", 5),
            new Motorcycle("Yamaha", "R1", false),
            new Car("Ford", "Focus", 4),
            new Motorcycle("Ducati", "Panigale", false)
        };

        foreach (var vehicle in vehicles)
        {
            VehicleService service = new VehicleService(vehicle);
            service.DisplayInfo();
            service.UseVehicle();
            Console.WriteLine(new string('-', 40));
        }

        Console.WriteLine();
        Console.WriteLine("=== Гибкость внедрения зависимостей ===");

        // Можно легко заменить реализацию без изменения кода сервиса
        IVehicle electricCar = new Car("Tesla", "Model S", 4);
        VehicleService teslaService = new VehicleService(electricCar);
        
        Console.WriteLine("Работа с электромобилем:");
        teslaService.DisplayInfo();
        teslaService.UseVehicle();
    }
}
````

````
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== Демонстрация внедрения зависимостей ===");

// Создание объектов
IAnimal myDog = new Dog("Рекс", 3, "Немецкая овчарка");
IAnimal myCat = new Cat("Мурка", 5, true);
IAnimal myBird = new Bird("Кеша", 2, 25.5);

// Создание сервисов с внедрением зависимостей
AnimalService dogService = new AnimalService(myDog);
AnimalService catService = new AnimalService(myCat);
AnimalService birdService = new AnimalService(myBird);

// Использование сервисов
Console.WriteLine("--- Работа с собакой ---");
dogService.DisplayInfo();
dogService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("--- Работа с кошкой ---");
catService.DisplayInfo();
catService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("--- Работа с птицей ---");
birdService.DisplayInfo();
birdService.CareForAnimal();

Console.WriteLine();
Console.WriteLine("=== Полиморфизм: коллекция животных ===");

// Коллекция объектов типа IAnimal
IAnimal[] animals = new IAnimal[]
{
    new Dog("Бобик", 2, "Дворняга"),
    new Cat("Сима", 4, false),
    new Bird("Попугай", 1, 15.0),
    new Dog("Шарик", 6, "Лабрадор")
};

foreach (var animal in animals)
{
    AnimalService service = new AnimalService(animal);
    service.DisplayInfo();
    service.CareForAnimal();
    Console.WriteLine(new string('-', 40));
}

Console.WriteLine();
Console.WriteLine("=== Гибкость внедрения зависимостей ===");

// Можно легко заменить реализацию без изменения кода сервиса
IAnimal exoticBird = new Bird("Ара", 10, 80.0);
AnimalService exoticService = new AnimalService(exoticBird);

Console.WriteLine("Работа с экзотической птицей:");
exoticService.DisplayInfo();
exoticService.CareForAnimal();

// --- КЛАССЫ И ИНТЕРФЕЙСЫ ---

// 1. Определение интерфейса IAnimal
public interface IAnimal
{
    void Eat();
    void Sleep();
    string GetName();
}

// 2. Абстрактный базовый класс Animal, реализующий IAnimal
public abstract class Animal : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    protected double energyLevel;

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        energyLevel = 100;
    }

    public virtual void Eat()
    {
        energyLevel += 20;
        Console.WriteLine($"{Name} поел. Уровень энергии: {energyLevel}");
    }

    public virtual void Sleep()
    {
        energyLevel = 100;
        Console.WriteLine($"{Name} поспал. Энергия восстановлена.");
    }

    public string GetName()
    {
        return Name;
    }

    public abstract void MakeSound();

    public virtual void Play()
    {
        Console.WriteLine($"{Name} играет.");
        energyLevel -= 10;
    }
}

// 3. Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} (собака) жадно ест корм.");
        base.Eat();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик.");
    }
}

// 4. Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void Sleep()
    {
        Console.WriteLine($"{Name} (кошка) спит на подоконнике.");
        base.Sleep();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу...");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет.");
    }
}

// 5. Производный класс Bird
public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void Eat()
    {
        Console.WriteLine($"{Name} клюет зерна.");
        base.Eat();
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит с размахом крыльев {WingSpan} см.");
    }
}

// 6. Класс сервиса, использующий внедрение зависимостей
public class AnimalService
{
    private readonly IAnimal _animal;

    public AnimalService(IAnimal animal)
    {
        _animal = animal;
    }

    public void CareForAnimal()
    {
        Console.WriteLine($"--- Уход за животным: {_animal.GetName()} ---");

        _animal.Eat();

        if (_animal is Dog dog)
        {
            dog.Fetch();
        }
        else if (_animal is Cat cat)
        {
            cat.Purr();
        }
        else if (_animal is Bird bird)
        {
            bird.Fly();
        }

        _animal.Sleep();
    }

    public void DisplayInfo()
    {
        if (_animal is Animal a)
        {
            Console.WriteLine($"Имя: {a.Name}, Возраст: {a.Age} лет");
        }
    }
}
````
