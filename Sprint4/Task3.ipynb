Создать базовый класс Invoice в C#, который будет представлять информацию о фактурах за поставленные товары или оказанные услуги. На основе этого класса разработать 2-3 производных класса, демонстрирующих принципы наследования и полиморфизма. В каждом из классов должны быть реализованы новые атрибуты и методы, а также переопределены некоторые методы базового класса для
демонстрации полиморфизма.
Требования к базовому классу Invoice:

• Атрибуты: Номер фактуры (InvoiceNumber), Дата выдачи (IssueDate), Общая
сумма (TotalAmount).

• Методы:

o CalculateTotal(): метод для расчета общей суммы по фактуре.

o AddLine(LineItem lineItem): метод для добавления позиции в фактуру.

o RemoveLine(LineItem lineItem): метод для удаления позиции из фактуры.

Требования к производным классам:

1. ТоварнаяФактура (GoodsInvoice): Должна содержать дополнительные атрибуты, такие как Дата поставки (SupplyDate). Метод AddLine() должен быть переопределен для добавления информации о дате поставки товара при добавлении позиции.

2. УслуговаяФактура (ServiceInvoice): Должна содержать дополнительные атрибуты, такие как Дата оказания услуги (ServiceDate). Метод RemoveLine() должен быть переопределен для добавления информации о причине аннулирования услуги при удалении позиции.

3. КомбинированнаяФактура (CombinedInvoice) (если требуется третий класс): Должна содержать дополнительные атрибуты, такие как Наличие возврата (ReturnAllowed). Метод CalculateTotal() должен быть переопределен для учета возможного возврата товара или услуги при расчете общей суммы.

```
using System;
using System.Collections.Generic;
using System.Linq;

// ========================
// 1. Интерфейс IInvoice
// ========================
public interface IInvoice
{
    void CalculateTotal();
    void AddLine(LineItem lineItem);
    void RemoveLine(LineItem lineItem);
    string GetInvoiceNumber();
    DateTime GetIssueDate();
    decimal GetTotalAmount();
}

// ========================
// 2. Базовый класс Invoice (абстрактный)
// ========================
public abstract class Invoice : IInvoice
{
    // Атрибуты из задания
    public string InvoiceNumber { get; set; }
    public DateTime IssueDate { get; set; }
    public decimal TotalAmount { get; protected set; }

    // Новые атрибуты (добавлено по требованию)
    public string CustomerName { get; set; }
    public string Currency { get; set; } = "RUB";
    public bool IsPaid { get; set; } = false;
    public List<LineItem> Lines { get; set; } = new List<LineItem>();

    // Конструктор
    public Invoice(string invoiceNumber, DateTime issueDate, string customerName)
    {
        InvoiceNumber = invoiceNumber;
        IssueDate = issueDate;
        CustomerName = customerName;
        TotalAmount = 0m;
    }

    // Явная реализация интерфейса
    void IInvoice.CalculateTotal()
    {
        CalculateTotalInternal();
    }

    void IInvoice.AddLine(LineItem lineItem)
    {
        AddLineInternal(lineItem);
    }

    void IInvoice.RemoveLine(LineItem lineItem)
    {
        RemoveLineInternal(lineItem);
    }

    string IInvoice.GetInvoiceNumber() => InvoiceNumber;
    DateTime IInvoice.GetIssueDate() => IssueDate;
    decimal IInvoice.GetTotalAmount() => TotalAmount;

    // Виртуальные методы для переопределения в наследниках
    protected virtual void CalculateTotalInternal()
    {
        TotalAmount = Lines.Sum(l => l.Quantity * l.Price);
        Console.WriteLine($"Общая сумма пересчитана: {TotalAmount} {Currency}");
    }

    protected virtual void AddLineInternal(LineItem lineItem)
    {
        Lines.Add(lineItem);
        CalculateTotalInternal();
        Console.WriteLine($"Позиция добавлена: {lineItem.Description} x{lineItem.Quantity}");
    }

    protected virtual void RemoveLineInternal(LineItem lineItem)
    {
        if (Lines.Contains(lineItem))
        {
            Lines.Remove(lineItem);
            CalculateTotalInternal();
            Console.WriteLine($"Позиция удалена: {lineItem.Description}");
        }
        else
        {
            Console.WriteLine("Позиция не найдена.");
        }
    }

    // Новые методы (добавлено по требованию)
    public virtual void PrintSummary()
    {
        Console.WriteLine($"\n=== ФАКТУРА ===");
        Console.WriteLine($"Номер: {InvoiceNumber}");
        Console.WriteLine($"Дата: {IssueDate:d}");
        Console.WriteLine($"Клиент: {CustomerName}");
        Console.WriteLine($"Валюта: {Currency}");
        Console.WriteLine($"Статус оплаты: {(IsPaid ? "Оплачено" : "Не оплачено")}");
        Console.WriteLine($"Позиций: {Lines.Count}");
        foreach (var line in Lines)
        {
            Console.WriteLine($"  - {line.Description}: {line.Quantity} шт. x {line.Price} {Currency} = {line.Quantity * line.Price} {Currency}");
        }
        Console.WriteLine($"ИТОГО: {TotalAmount} {Currency}\n");
    }

    public virtual void MarkAsPaid()
    {
        IsPaid = true;
        Console.WriteLine($"Фактура {InvoiceNumber} помечена как оплаченная.");
    }

    public virtual void ApplyDiscount(decimal percent)
    {
        if (percent > 0 && percent <= 100)
        {
            var discount = TotalAmount * (percent / 100);
            TotalAmount -= discount;
            Console.WriteLine($"Применена скидка {percent}%: -{discount} {Currency}. Итого: {TotalAmount} {Currency}");
        }
        else
        {
            Console.WriteLine("Некорректный процент скидки.");
        }
    }
}

// ========================
// 3. Производный класс GoodsInvoice (Товарная фактура)
// ========================
public class GoodsInvoice : Invoice
{
    // Дополнительный атрибут из задания
    public DateTime SupplyDate { get; set; }

    // Новые атрибуты (добавлено по требованию)
    public string SupplierName { get; set; }
    public string WarehouseLocation { get; set; }
    public int DeliveryDays { get; set; }

    public GoodsInvoice(string invoiceNumber, DateTime issueDate, string customerName, DateTime supplyDate)
        : base(invoiceNumber, issueDate, customerName)
    {
        SupplyDate = supplyDate;
    }

    // Переопределение метода AddLine (из задания)
    protected override void AddLineInternal(LineItem lineItem)
    {
        // Добавляем информацию о дате поставки товара при добавлении позиции
        lineItem.Description += $" [Поставка: {SupplyDate:d}]";
        base.AddLineInternal(lineItem);
        Console.WriteLine($"[GoodsInvoice] Товар будет поставлен: {SupplyDate:d}");
    }

    // Новые методы (добавлено по требованию)
    public void SetDeliverySchedule(int days)
    {
        DeliveryDays = days;
        Console.WriteLine($"Срок доставки установлен: {days} дней.");
    }

    public void UpdateSupplier(string supplierName, string warehouse)
    {
        SupplierName = supplierName;
        WarehouseLocation = warehouse;
        Console.WriteLine($"Поставщик обновлен: {supplierName}, Склад: {warehouse}");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"[ТОВАРНАЯ ФАКТУРА]");
        Console.WriteLine($"Дата поставки: {SupplyDate:d}");
        Console.WriteLine($"Поставщик: {SupplierName ?? "Не указан"}");
        Console.WriteLine($"Склад: {WarehouseLocation ?? "Не указан"}");
        Console.WriteLine($"Срок доставки: {DeliveryDays} дн.");
        Console.WriteLine(new string('-', 40));
    }
}

// ========================
// 4. Производный класс ServiceInvoice (Услуговая фактура)
// ========================
public class ServiceInvoice : Invoice
{
    // Дополнительный атрибут из задания
    public DateTime ServiceDate { get; set; }

    // Новые атрибуты (добавлено по требованию)
    public string ServiceType { get; set; }
    public string TechnicianName { get; set; }
    public TimeSpan EstimatedDuration { get; set; }

    public ServiceInvoice(string invoiceNumber, DateTime issueDate, string customerName, DateTime serviceDate)
        : base(invoiceNumber, issueDate, customerName)
    {
        ServiceDate = serviceDate;
    }

    // Переопределение метода RemoveLine (из задания)
    protected override void RemoveLineInternal(LineItem lineItem)
    {
        // Добавляем информацию о причине аннулирования услуги при удалении позиции
        Console.WriteLine($"[ServiceInvoice] Услуга '{lineItem.Description}' аннулирована. Причина: отмена заказа клиента.");
        base.RemoveLineInternal(lineItem);
    }

    // Новые методы (добавлено по требованию)
    public void AssignTechnician(string name)
    {
        TechnicianName = name;
        Console.WriteLine($"Назначен техник: {name}");
    }

    public void SetServiceDuration(TimeSpan duration)
    {
        EstimatedDuration = duration;
        Console.WriteLine($"Ожидаемая длительность услуги: {duration.Hours} ч {duration.Minutes} мин");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"[УСЛУГОВАЯ ФАКТУРА]");
        Console.WriteLine($"Дата оказания услуги: {ServiceDate:d}");
        Console.WriteLine($"Тип услуги: {ServiceType ?? "Не указан"}");
        Console.WriteLine($"Техник: {TechnicianName ?? "Не назначен"}");
        Console.WriteLine($"Длительность: {EstimatedDuration.Hours} ч {EstimatedDuration.Minutes} мин");
        Console.WriteLine(new string('-', 40));
    }
}

// ========================
// 5. Производный класс CombinedInvoice (Комбинированная фактура)
// ========================
public class CombinedInvoice : Invoice
{
    // Дополнительный атрибут из задания
    public bool ReturnAllowed { get; set; }

    // Новые атрибуты (добавлено по требованию)
    public DateTime ReturnDeadline { get; set; }
    public decimal ReturnPenaltyPercent { get; set; }
    public string Notes { get; set; }

    public CombinedInvoice(string invoiceNumber, DateTime issueDate, string customerName, bool returnAllowed)
        : base(invoiceNumber, issueDate, customerName)
    {
        ReturnAllowed = returnAllowed;
        ReturnDeadline = issueDate.AddDays(30); // По умолчанию 30 дней
        ReturnPenaltyPercent = 10; // По умолчанию 10%
    }

    // Переопределение метода CalculateTotal (из задания)
    protected override void CalculateTotalInternal()
    {
        base.CalculateTotalInternal();
        if (ReturnAllowed)
        {
            Console.WriteLine("[CombinedInvoice] Учтена возможность возврата. При возврате может быть применена пеня.");
        }
    }

    // Новые методы (добавлено по требованию)
    public void SetReturnPolicy(DateTime deadline, decimal penaltyPercent)
    {
        ReturnDeadline = deadline;
        ReturnPenaltyPercent = penaltyPercent;
        Console.WriteLine($"Политика возврата установлена: до {deadline:d}, пеня {penaltyPercent}%");
    }

    public void AddNote(string note)
    {
        Notes = note;
        Console.WriteLine($"Добавлена заметка: {note}");
    }

    public override void PrintSummary()
    {
        base.PrintSummary();
        Console.WriteLine($"[КОМБИНИРОВАННАЯ ФАКТУРА]");
        Console.WriteLine($"Возврат разрешен: {(ReturnAllowed ? "Да" : "Нет")}");
        Console.WriteLine($"Срок возврата: {ReturnDeadline:d}");
        Console.WriteLine($"Пеня за возврат: {ReturnPenaltyPercent}%");
        Console.WriteLine($"Заметки: {Notes ?? "Отсутствуют"}");
        Console.WriteLine(new string('-', 40));
    }
}

// ========================
// 6. Класс LineItem (позиция в фактуре)
// ========================
public class LineItem
{
    public string Description { get; set; }
    public int Quantity { get; set; }
    public decimal Price { get; set; }

    public LineItem(string description, int quantity, decimal price)
    {
        Description = description;
        Quantity = quantity;
        Price = price;
    }
}

// ========================
// 7. Сервис InvoiceService с внедрением зависимостей
// ========================
public class InvoiceService
{
    private readonly IInvoice _invoice;

    // Внедрение зависимости через конструктор
    public InvoiceService(IInvoice invoice)
    {
        _invoice = invoice;
    }

    public void ProcessInvoice()
    {
        Console.WriteLine($"\n=== ОБРАБОТКА ФАКТУРЫ {_invoice.GetInvoiceNumber()} ===");

        // Вызов методов через интерфейс
        _invoice.CalculateTotal();

        // Если это конкретный тип — можно вызвать специфичные методы
        if (_invoice is GoodsInvoice goods)
        {
            goods.SetDeliverySchedule(7);
            goods.UpdateSupplier("ООО 'Торгснаб'", "Склад №3");
        }
        else if (_invoice is ServiceInvoice service)
        {
            service.AssignTechnician("Иван Петров");
            service.SetServiceDuration(TimeSpan.FromHours(2.5));
        }
        else if (_invoice is CombinedInvoice combined)
        {
            combined.SetReturnPolicy(DateTime.Now.AddDays(60), 5);
            combined.AddNote("Клиент запросил расширенную гарантию.");
        }

        // Печать сводки
        if (_invoice is Invoice inv)
        {
            inv.PrintSummary();
        }

        // Применение скидки
        _invoice.ApplyDiscount(5); // Это вызовет ошибку, т.к. ApplyDiscount не в интерфейсе!
        // Исправление: нужно привести к Invoice или добавить метод в интерфейс
        // Лучше сделать так:
        if (_invoice is Invoice concreteInvoice)
        {
            concreteInvoice.ApplyDiscount(5);
        }

        // Пометка как оплаченной
        if (_invoice is Invoice paidInvoice)
        {
            paidInvoice.MarkAsPaid();
        }
    }
}

// ========================
// 8. Программа для демонстрации
// ========================
class Program
{
    static void Main()
    {
        Console.WriteLine("=== ДЕМОНСТРАЦИЯ РАБОТЫ С ФАКТУРАМИ ===\n");

        // Создание объектов
        var goodsInvoice = new GoodsInvoice(
            "INV-001", 
            DateTime.Today, 
            "ООО 'Ромашка'", 
            DateTime.Today.AddDays(5)
        );
        goodsInvoice.Lines.Add(new LineItem("Ноутбук ASUS", 2, 45000));
        goodsInvoice.Lines.Add(new LineItem("Мышь беспроводная", 5, 1500));

        var serviceInvoice = new ServiceInvoice(
            "INV-002", 
            DateTime.Today, 
            "ИП Иванов", 
            DateTime.Today.AddDays(2)
        );
        serviceInvoice.ServiceType = "Ремонт компьютера";
        serviceInvoice.Lines.Add(new LineItem("Диагностика", 1, 2000));
        serviceInvoice.Lines.Add(new LineItem("Замена HDD", 1, 5000));

        var combinedInvoice = new CombinedInvoice(
            "INV-003", 
            DateTime.Today, 
            "ЗАО 'Вектор'", 
            true
        );
        combinedInvoice.Lines.Add(new LineItem("Консультация", 3, 3000));
        combinedInvoice.Lines.Add(new LineItem("Программное обеспечение", 1, 15000));

        // Создание сервисов с внедрением зависимостей
        var goodsService = new InvoiceService(goodsInvoice);
        var serviceService = new InvoiceService(serviceInvoice);
        var combinedService = new InvoiceService(combinedInvoice);

        // Обработка фактур
        goodsService.ProcessInvoice();
        serviceService.ProcessInvoice();
        combinedService.ProcessInvoice();

        Console.WriteLine("\n=== ПОЛИМОРФИЗМ: КОЛЛЕКЦИЯ ФАКТУР ===");

        IInvoice[] invoices = { goodsInvoice, serviceInvoice, combinedInvoice };

        foreach (var invoice in invoices)
        {
            var svc = new InvoiceService(invoice);
            svc.ProcessInvoice();
            Console.WriteLine(new string('=', 60));
        }
    }
}
```
