<h1 style="color:DodgerBlue">Явная реализация интерфейса</h1>

В C# интерфейсы могут быть реализованы явным и неявным образом. Давайте подробно рассмотрим, что такое явная реализация интерфейсов, реализация интерфейсов в базовых и производных классах, а также наследование интерфейсов, используя примеры классов `Vehicle`, `Car` и `Motorcycle`.

### Явная реализация интерфейсов

Явная реализация интерфейса используется для того, чтобы скрыть методы интерфейса от стандартного контекста класса. Это означает, что методы интерфейса могут быть вызваны только через ссылку на интерфейс, а не через экземпляр класса.

#### Пример явной реализации

Рассмотрим интерфейс `IVehicle` и его явную реализацию в классе `Car`.

```csharp
public interface IVehicle
{
    void Start();
    void Stop();
    int GetCurrentSpeed();
}

public class Car : IVehicle
{
    public string Make { get; set; }
    public string Model { get; set; }
    private int currentSpeed;

    // Явная реализация интерфейса
    void IVehicle.Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} запущен.");
    }

    void IVehicle.Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} остановлен.");
    }

    // Неявная реализация метода
    public int GetCurrentSpeed()
    {
        return currentSpeed;
    }
}
```

В этом примере методы `Start` и `Stop` реализованы явно, и для их вызова необходимо использовать объект типа `IVehicle`:

```csharp

IVehicle myCar = new Car { Make = "Toyota", Model = "Camry" };
myCar.Start(); // Корректно, вызывает явную реализацию
// myCar.Stop(); // Корректно, вызывает явную реализацию
```

Если вы попытаетесь вызвать `Start` или `Stop` непосредственно через экземпляр `Car`, это приведёт к ошибке компиляции:

```csharp
Car myCar = new Car();
myCar.Start(); // Ошибка компиляции
```

### Реализация интерфейсов в базовых и производных классах

Интерфейсы могут быть реализованы в базовых классах и наследоваться в производных классах. Давайте рассмотрим эту концепцию на примере, включающем классы `Vehicle`, `Car` и `Motorcycle`.

#### Базовый класс Vehicle

```csharp
public abstract class Vehicle : IVehicle
{
    public string Make { get; set; }
    public string Model { get; set; }
    protected int currentSpeed;

    public abstract void Start();  // Оставляем абстрактным
    public abstract void Stop();   // Оставляем абстрактным

    public virtual int GetCurrentSpeed()
    {
        return currentSpeed;
    }
}
```

#### Реализация в производных классах

Теперь реализуем `Car` и `Motorcycle`.

```csharp
public class Car : Vehicle
{
    public override void Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} запущен.");
    }

    public override void Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} остановлен.");
    }
}

public class Motorcycle : Vehicle
{
    public override void Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл запущен.");
    }

    public override void Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл остановлен.");
    }
}
```

Теперь оба класса `Car` и `Motorcycle` обязаны реализовать методы `Start` и `Stop`, которые определены в абстрактном классе `Vehicle`.

### Наследование интерфейсов

Интерфейсы могут наследовать друг от друга. Это позволяет создавать сложные иерархии интерфейсов. Например, мы можем создать интерфейс `IMotorizedVehicle`, который наследует `IVehicle`.

```csharp
public interface IMotorizedVehicle : IVehicle
{
    void RevEngine();
}

public class Motorcycle : Vehicle, IMotorizedVehicle
{
    public override void Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл запущен.");
    }

    public override void Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл остановлен.");
    }

    public void RevEngine()
    {
        Console.WriteLine("Мотоцикл рычит!");
    }
}
```

Таким образом, `Motorcycle` теперь может реализовать все методы из обоих интерфейсов `IVehicle` и `IMotorizedVehicle`.

### Заключение

1. **Явная реализация** интерфейсов скрывает функциональность классов от общего доступа, что помогает избежать путаницы с методами.
   
2. **Реализация интерфейсов в базовых и производных классах** позволяет создавать иерархию классов и обеспечивает обязательность реализации методов в производных классах.

3. **Наследование интерфейсов** позволяет комбинировать несколько интерфейсов, создавая более сложные структуры, что делает код более гибким и расширяемым.

Эти механизмы являются важными инструментами в инструменте разработчика, позволяя создавать более структурированные и управляемые программы.

````
using System;

// 1. Базовый интерфейс
public interface IVehicle
{
    void Start();
    void Stop();
    int GetCurrentSpeed();
}

// 2. Расширенный интерфейс, наследующий IVehicle
public interface IMotorizedVehicle : IVehicle
{
    void RevEngine();
}

// 3. Абстрактный базовый класс, реализующий IVehicle
public abstract class Vehicle : IVehicle
{
    public string Make { get; set; }
    public string Model { get; set; }
    protected int currentSpeed;

    // Конструктор для инициализации свойств
    public Vehicle(string make, string model)
    {
        Make = make;
        Model = model;
        currentSpeed = 0;
    }

    // Абстрактные методы - должны быть реализованы в производных классах
    public abstract void Start();
    public abstract void Stop();

    // Виртуальный метод с базовой реализацией
    public virtual int GetCurrentSpeed()
    {
        return currentSpeed;
    }
}

// 4. Производный класс Car с ЯВНОЙ реализацией интерфейса IVehicle
// Это означает, что методы Start() и Stop() доступны только через ссылку типа IVehicle
public class Car : Vehicle, IVehicle
{
    public int NumberOfDoors { get; set; }

    public Car(string make, string model, int numberOfDoors) : base(make, model)
    {
        NumberOfDoors = numberOfDoors;
    }

    // Явная реализация интерфейса IVehicle
    void IVehicle.Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} запущен (явная реализация).");
    }

    void IVehicle.Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} остановлен (явная реализация).");
    }

    // Неявная реализация метода GetCurrentSpeed() - доступна напрямую
    public override int GetCurrentSpeed()
    {
        return currentSpeed;
    }

    // Реализация абстрактных методов из базового класса Vehicle
    // Эти методы будут вызываться при обращении к объекту как к Car
    public override void Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} запущен (прямой вызов).");
    }

    public override void Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} остановлен (прямой вызов).");
    }

    // Уникальный метод для автомобиля
    public void Accelerate(int amount)
    {
        currentSpeed += amount;
        Console.WriteLine($"Автомобиль разгоняется до {currentSpeed} км/ч.");
    }
}

// 5. Производный класс Motorcycle, реализующий расширенный интерфейс IMotorizedVehicle
public class Motorcycle : Vehicle, IMotorizedVehicle
{
    public bool HasSidecar { get; set; }

    public Motorcycle(string make, string model, bool hasSidecar) : base(make, model)
    {
        HasSidecar = hasSidecar;
    }

    // Реализация абстрактных методов из базового класса Vehicle
    public override void Start()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл запущен.");
    }

    public override void Stop()
    {
        currentSpeed = 0;
        Console.WriteLine($"{Make} {Model} мотоцикл остановлен.");
    }

    // Реализация метода из интерфейса IMotorizedVehicle
    public void RevEngine()
    {
        Console.WriteLine("Мотоцикл рычит!");
    }
}

// 6. Пример использования всех классов и интерфейсов
class Program
{
    static void Main()
    {
        Console.WriteLine("=== Демонстрация явной и неявной реализации ===");

        // Создание экземпляра Car
        Car myCar = new Car("Toyota", "Camry", 4);

        // Прямой вызов методов (используется реализация из класса Car)
        myCar.Start();          // Вызовет: "Toyota Camry запущен (прямой вызов)."
        myCar.Accelerate(50);
        Console.WriteLine($"Скорость: {myCar.GetCurrentSpeed()} км/ч");
        myCar.Stop();           // Вызовет: "Toyota Camry остановлен (прямой вызов)."

        Console.WriteLine();

        // Вызов через интерфейс IVehicle (используется явная реализация)
        IVehicle vehicleInterface = myCar;
        vehicleInterface.Start();   // Вызовет: "Toyota Camry запущен (явная реализация)."
        vehicleInterface.Stop();    // Вызовет: "Toyota Camry остановлен (явная реализация)."
        // vehicleInterface.Accelerate(10); // Ошибка компиляции! Метод не определен в интерфейсе

        Console.WriteLine();
        Console.WriteLine("=== Демонстрация наследования интерфейсов ===");

        // Создание экземпляра Motorcycle
        Motorcycle myMotorcycle = new Motorcycle("Harley-Davidson", "Street", false);

        // Работа через интерфейс IMotorizedVehicle
        IMotorizedVehicle motorizedVehicle = myMotorcycle;
        motorizedVehicle.Start();       // Из интерфейса IVehicle
        motorizedVehicle.RevEngine();   // Из интерфейса IMotorizedVehicle
        motorizedVehicle.Stop();        // Из интерфейса IVehicle

        Console.WriteLine();
        Console.WriteLine("=== Полиморфизм: коллекция транспортных средств ===");

        // Коллекция объектов типа IVehicle
        IVehicle[] vehicles = new IVehicle[]
        {
            new Car("BMW", "X5", 5),
            new Motorcycle("Yamaha", "R1", false),
            new Car("Ford", "Focus", 4)
        };

        foreach (var vehicle in vehicles)
        {
            vehicle.Start();
            // Можно проверить тип и вызвать специфичные методы
            if (vehicle is Car car)
            {
                car.Accelerate(30);
            }
            else if (vehicle is Motorcycle motorcycle)
            {
                motorcycle.RevEngine();
            }
            vehicle.Stop();
            Console.WriteLine(new string('-', 30));
        }
    }
}
````

````
using System;

// --- ИСПОЛНЯЕМАЯ ЧАСТЬ ---

Console.WriteLine("=== Демонстрация явной реализации интерфейса ===");

// Создание экземпляра Dog
Dog myDog = new Dog("Рекс", 3, "Немецкая овчарка");

// Прямой вызов методов (используется реализация из класса Dog)
myDog.MakeSound();      // Вызовет: "Рекс лает: Гав-гав!"
myDog.Play();           // Вызовет: "Рекс играет."
myDog.Fetch();          // Вызовет: "Рекс приносит мячик."

Console.WriteLine();

// Вызов через интерфейс IAnimal (используется явная реализация)
IAnimal animalInterface = myDog;
animalInterface.Eat();      // Вызовет: "Рекс поел. Уровень энергии: 120"
animalInterface.Sleep();    // Вызовет: "Рекс поспал. Энергия восстановлена."
string name = animalInterface.GetName();
Console.WriteLine($"Имя животного: {name}");

// animalInterface.MakeSound(); // Ошибка компиляции! Метод не определен в интерфейсе
// animalInterface.Fetch();     // Ошибка компиляции! Метод не определен в интерфейсе

Console.WriteLine();
Console.WriteLine("=== Демонстрация с другими животными ===");

// Создание экземпляра Cat
Cat myCat = new Cat("Мурка", 5, true);

// Прямой вызов
myCat.MakeSound();  // "Мурка мяукает: Мяу..."
myCat.Purr();       // "Мурка мурлычет."

// Вызов через интерфейс
IAnimal catInterface = myCat;
catInterface.Eat();
catInterface.Sleep();

Console.WriteLine();

// Создание экземпляра Bird
Bird myBird = new Bird("Кеша", 2, 25.5);

// Прямой вызов
myBird.MakeSound(); // "Кеша чирикает: Чик-чирик!"
myBird.Fly();       // "Кеша летит с размахом крыльев 25.5 см."

// Вызов через интерфейс
IAnimal birdInterface = myBird;
birdInterface.Eat();
birdInterface.Sleep();

Console.WriteLine();
Console.WriteLine("=== Полиморфизм: коллекция животных ===");

// Коллекция объектов типа IAnimal
IAnimal[] animals = new IAnimal[]
{
    new Dog("Бобик", 2, "Дворняга"),
    new Cat("Сима", 4, false),
    new Bird("Попугай", 1, 15.0)
};

foreach (var animal in animals)
{
    animal.Eat();
    animal.Sleep();
    Console.WriteLine(new string('-', 30));
}

// Чтобы вызвать уникальные методы, нужно привести тип
Console.WriteLine("\n=== Вызов уникальных методов после приведения типа ===");
foreach (var animal in animals)
{
    if (animal is Dog dog)
    {
        dog.Fetch();
    }
    else if (animal is Cat cat)
    {
        cat.Purr();
    }
    else if (animal is Bird bird)
    {
        bird.Fly();
    }
}

// --- КЛАССЫ И ИНТЕРФЕЙСЫ ---

// 1. Определение интерфейса IAnimal
public interface IAnimal
{
    void Eat();
    void Sleep();
    string GetName();
}

// 2. Абстрактный базовый класс Animal, реализующий IAnimal
public abstract class Animal : IAnimal
{
    public string Name { get; set; }
    public int Age { get; set; }
    protected double energyLevel;

    public Animal(string name, int age)
    {
        Name = name;
        Age = age;
        energyLevel = 100;
    }

    // Явная реализация методов интерфейса IAnimal
    void IAnimal.Eat()
    {
        energyLevel += 20;
        Console.WriteLine($"{Name} поел. Уровень энергии: {energyLevel}");
    }

    void IAnimal.Sleep()
    {
        energyLevel = 100;
        Console.WriteLine($"{Name} поспал. Энергия восстановлена.");
    }

    string IAnimal.GetName()
    {
        return Name;
    }

    // Абстрактный метод, который должны реализовать все наследники
    public abstract void MakeSound();

    // Неявный метод, доступный напрямую
    public virtual void Play()
    {
        Console.WriteLine($"{Name} играет.");
        energyLevel -= 10;
    }
}

// 3. Производный класс Dog
public class Dog : Animal
{
    public string Breed { get; set; }

    public Dog(string name, int age, string breed) : base(name, age)
    {
        Breed = breed;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} лает: Гав-гав!");
    }

    public void Fetch()
    {
        Console.WriteLine($"{Name} приносит мячик.");
    }
}

// 4. Производный класс Cat
public class Cat : Animal
{
    public bool IsIndoor { get; set; }

    public Cat(string name, int age, bool isIndoor) : base(name, age)
    {
        IsIndoor = isIndoor;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} мяукает: Мяу...");
    }

    public void Purr()
    {
        Console.WriteLine($"{Name} мурлычет.");
    }
}

// 5. Производный класс Bird
public class Bird : Animal
{
    public double WingSpan { get; set; }

    public Bird(string name, int age, double wingSpan) : base(name, age)
    {
        WingSpan = wingSpan;
    }

    public override void MakeSound()
    {
        Console.WriteLine($"{Name} чирикает: Чик-чирик!");
    }

    public void Fly()
    {
        Console.WriteLine($"{Name} летит с размахом крыльев {WingSpan} см.");
    }
}
````
